```python
# ============================================================================
# COMPREHENSIVE EXOPLANET SPECTROSCOPY & ASTROBIOLOGY ANALYSIS FRAMEWORK
# ============================================================================
#
# MERGED EDITION: v10.0 (Combined Real Data + Asymmetric Rotor Theory)
# Combines: exoplanet_discovery_v7-8-6.3_2_5_3.txt + exoplanet_discovery_v4.txt
#
# Features:
# 1. Asymmetric Rotor Theory (Ray's κ parameter)
# 2. Comprehensive Molecular Database with Real Data from extract_thesis2.pdf
# 3. TOSH Theory Implementation for Anharmonic Corrections
# 4. Exoplanet Discovery Data Integration
# 5. Theoretical Astrophysics Models
# 6. Quantum Computing Framework (Optional)
# 7. Biosignature Detection Systems
# 8. Enhanced Visualization Tools
#
# Author: Merged Analysis System
# Version: 10.0 (Merged Edition)
# Date: 2025-02-09
# Platform: Google Colab
# ============================================================================

# ============================================================================
# SECTION 1: INSTALL ALL DEPENDENCIES
# ============================================================================
print("="*70)
print("INSTALLING COMPREHENSIVE DEPENDENCIES FOR MERGED FRAMEWORK")
print("="*70)

# First, upgrade pip and setuptools
!pip install -q --upgrade pip setuptools wheel
!pip install -q jedi>=0.16  # Fix IPython dependency

# Core packages
!pip install -q numpy scipy matplotlib seaborn pandas
!pip install -q requests beautifulsoup4 lxml
!pip install -q plotly sympy

# Astronomy and exoplanet packages
!pip install -q astropy astroquery
!pip install -q pubchempy

# Machine Learning packages (optional)
!pip install -q scikit-learn
!pip install -q torch torchvision torchaudio --index-url https://download.pytorch.org/whl/cpu
!pip install -q transformers

# Chemistry packages (optional)
!pip install -q rdkit-pypi pyscf

# Optional: Quantum Computing (can be skipped if not needed)
try:
    !pip install -q 'qiskit>=1.0' 'qiskit-aer>=0.14' 'qiskit-algorithms' 'qiskit-machine-learning' 'qiskit-nature' 'qiskit-optimization' 'qiskit-ibm-runtime'
    print("✓ Quantum computing packages installed")
except:
    print("⚠ Quantum computing packages installation skipped or failed")

print("\n✓ Checking core installations...")
!python -c "import numpy as np; print(f'NumPy version: {np.__version__}')"
!python -c "import pandas as pd; print(f'Pandas version: {pd.__version__}')"

# ============================================================================
# SECTION 2: IMPORT ALL LIBRARIES
# ============================================================================
print("\n" + "="*70)
print("IMPORTING ALL LIBRARIES")
print("="*70)

import sys
import warnings
warnings.filterwarnings('ignore')

# Core libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats, optimize, interpolate, special, linalg
import json
from datetime import datetime
import hashlib
import pickle
import os
import sympy as sp
from typing import Dict, List, Tuple, Optional, Any, Union
from itertools import product, combinations
from collections import defaultdict

# Astronomy and units
from astropy import units as u
from astropy import constants as const

# Visualization
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots

# Machine Learning and AI (optional)
try:
    import torch
    import torch.nn as nn
    import torch.optim as optim
    from sklearn.preprocessing import StandardScaler, MinMaxScaler
    from sklearn.decomposition import PCA
    from sklearn.manifold import TSNE
    from sklearn.cluster import KMeans
    ML_AVAILABLE = True
    print("✓ Machine learning libraries imported")
except ImportError as e:
    ML_AVAILABLE = False
    print(f"⚠ Machine learning libraries not available: {e}")

# Quantum Computing (optional)
try:
    from qiskit import QuantumCircuit, transpile, QuantumRegister, ClassicalRegister
    from qiskit_aer import AerSimulator, Aer
    from qiskit.visualization import plot_histogram, plot_bloch_multivector
    from qiskit.primitives import Sampler, Estimator
    from qiskit_algorithms import VQE, QAOA
    from qiskit_algorithms.optimizers import COBYLA, SLSQP, ADAM
    from qiskit_machine_learning.kernels import FidelityQuantumKernel
    from qiskit_machine_learning.algorithms import VQC, QSVC
    from qiskit.circuit.library import ZZFeatureMap, RealAmplitudes, EfficientSU2, PauliFeatureMap
    QISKIT_AVAILABLE = True
    print("✓ Quantum computing libraries imported")
except ImportError as e:
    QISKIT_AVAILABLE = False
    print(f"⚠ Quantum computing libraries not available: {e}")

# Try to import astronomy packages
try:
    from astroquery.simbad import Simbad
    from astroquery.nasa_exoplanet_archive import NasaExoplanetArchive
    ASTRO_AVAILABLE = True
    print("✓ Astroquery available")
except ImportError as e:
    ASTRO_AVAILABLE = False
    print(f"Note: astroquery not available: {e}")

# Try to import chemistry packages
try:
    import pubchempy as pcp
    PUBCHEM_AVAILABLE = True
    print("✓ PubChemPy available")
except ImportError as e:
    PUBCHEM_AVAILABLE = False
    print(f"Note: pubchempy not available: {e}")

# Configure plotting
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 10

print("\n" + "="*70)
print("LIBRARY IMPORT COMPLETE")
print("="*70)

# ============================================================================
# SECTION 3: ASYMMETRIC ROTOR THEORY
# ============================================================================
print("\n" + "="*70)
print("IMPLEMENTING ASYMMETRIC ROTOR THEORY")
print("="*70)

class AsymmetricRotorTheory:
    """Implementation of asymmetric rotor theory from Pandora spectral modeling"""

    def __init__(self):
        # Physical constants
        self.h = 6.62607015e-34        # Planck constant (J·s)
        self.hbar = self.h / (2 * np.pi)  # Reduced Planck constant (J·s)
        self.c = 299792458             # Speed of light (m/s)
        self.c_cm = self.c * 100       # Speed of light in cm/s
        self.k_B = 1.380649e-23        # Boltzmann constant (J/K)
        self.amu_to_kg = 1.66053906660e-27  # 1 amu in kg

        print("✓ Asymmetric Rotor Theory initialized")

    def calculate_ray_asymmetry_parameter(self, A: float, B: float, C: float) -> float:
        """
        Calculate Ray's asymmetry parameter κ (Eq. 5.15)

        κ = (2B - A - C) / (A - C)
        """
        if abs(A - C) < 1e-10:
            return 0.0  # Avoid division by zero

        kappa = (2 * B - A - C) / (A - C)
        return kappa

    def classify_asymmetric_top(self, kappa: float) -> str:
        """Classify asymmetric top based on κ value"""
        if kappa <= -0.9:
            return "Near-prolate symmetric top"
        elif kappa >= 0.9:
            return "Near-oblate symmetric top"
        elif -0.1 <= kappa <= 0.1:
            return "Maximum asymmetry (κ ≈ 0)"
        elif kappa < 0:
            return "Prolate-like asymmetric top"
        else:
            return "Oblate-like asymmetric top"

    def calculate_energy_levels(self, J_max: int, A: float, B: float, C: float) -> Dict[int, List[float]]:
        """
        Calculate asymmetric rotor energy levels using simplified method
        """
        energy_levels = {}

        for J in range(J_max + 1):
            levels = []

            # Simplified energy calculation for asymmetric top
            for K in range(-J, J + 1):
                # Simplified energy: E = B_eff * J(J+1) + (A-C) * K²
                B_eff = (A + B + C) / 3
                energy = B_eff * J * (J + 1) + (A - C) * K**2 / 100
                levels.append(energy)

            # Sort energy levels
            levels.sort()
            energy_levels[J] = levels

        return energy_levels

# Initialize theory
asymmetric_theory = AsymmetricRotorTheory()

# ============================================================================
# SECTION 4: COMPREHENSIVE MOLECULAR DATABASE (MERGED REAL DATA)
# ============================================================================
print("\n" + "="*70)
print("INITIALIZING COMPREHENSIVE MOLECULAR DATABASE")
print("(Merging real data from extract_thesis2.pdf with polyatomic molecules)")
print("="*70)

class ComprehensiveMolecularDatabase:
    """Complete molecular database with real data and asymmetric rotor capabilities"""

    def __init__(self):
        self.asymmetric_theory = AsymmetricRotorTheory()

        # Constants for unit conversions
        self.hartree_to_cm = 219474.63136320  # Conversion factor
        self.bohr_to_cm = 0.52917721067e-8    # Bohr radius in cm
        self.electron_mass_kg = 9.10938356e-31

        # REAL DIATOMIC DATA from extract_thesis2.pdf (Table 3.1)
        self.diatomic_molecules = {
            'H2': {
                'name': 'Diatomic Hydrogen',
                'formula': 'H₂',
                'molecular_weight': 2.016,
                'type': 'diatomic',
                'rotor_type': 'linear',
                'symmetry': 'homonuclear',
                'astrophysical_context': (
                    "Most abundant molecule in the Universe by several orders of magnitude. "
                    "Difficult to observe directly due to lack of permanent electric dipole moment. "
                    "Most transitions are quadrupolar and extremely weak."
                ),
                'biogenic_context': (
                    "On Earth, many microorganisms produce hydrogen as primary product through "
                    "metabolism under anaerobic conditions. Some microorganisms also consume H₂."
                ),
                'abiotic_sources': [
                    "Serpentinization of ferrous materials",
                    "Water splitting due to radioactive decay"
                ],
                'detection_methods': [
                    "Indirect techniques using gas-to-dust ratio",
                    "Quadrupole transitions (weak)",
                    "Emission in selected environments"
                ],
                'spectral_activity': 'Raman active, IR inactive',
                'nuclear_spin': 'Non-zero (alternating intensities)',
                'anharmonic_constants': {
                    'η_ii': 4.040e-4,      # [E_h][m_e]⁻¹[a_0]⁻²
                    'η_iii': -4.971e-5,    # [E_h][m_e]⁻³/²[a_0]⁻³
                    'η_iiii': 5.317e-6     # [E_h][m_e]⁻²[a_0]⁻⁴
                },
                'spectroscopic_constants': {
                    'band_origin': {
                        'literature': 4161.17,   # Campargue et al. (2012)
                        'tosh': 4174.71,
                        'harmonic': 4411.45,
                        'error_tosh': 13.54,
                        'error_harmonic': 250.28
                    },
                    'rotational_constants': {
                        'B0': {'literature': 59.3329, 'harmonic': 60.9513, 'tosh': 54.2350},
                        'B1': {'literature': 56.3732, 'harmonic': 60.9513, 'tosh': 54.2350}
                    },
                    'reduced_mass': 0.50391,           # amu
                    'equilibrium_length': 0.7414e-8,   # cm
                    'force_constant': 5.749e5,         # N/m
                    'branches_present': ['O', 'Q', 'S'],
                    'max_quantum_number': 20
                },
                'references': [
                    "Campargue et al. (2012)",
                    "Schwartz and Le Roy (1987)",
                    "Wakelam et al. (2017)"
                ]
            },

            'O2': {
                'name': 'Diatomic Oxygen',
                'formula': 'O₂',
                'molecular_weight': 32.00,
                'type': 'diatomic',
                'rotor_type': 'linear',
                'symmetry': 'homonuclear',
                'astrophysical_context': (
                    "Third most abundant element but molecular oxygen is elusive. "
                    "High chemical reactivity and lack of electric dipole moment. "
                    "Freeze-out onto dust grains limits gas-phase detection."
                ),
                'biogenic_context': (
                    "Crucial for most life on Earth with oxygenic photosynthesis as dominant "
                    "producing metabolism. Potential biosignature but also false positive."
                ),
                'abiotic_sources': [
                    "Runaway greenhouse effects",
                    "CO₂ photolysis on planets around M-class stars"
                ],
                'detection_methods': [
                    "Raman spectroscopy",
                    "Weak magnetic dipole transitions"
                ],
                'spectral_activity': 'Raman active, IR inactive',
                'nuclear_spin': 'Zero (alternating intensities absent)',
                'anharmonic_constants': {
                    'η_ii': 5.183e-5,
                    'η_iii': -1.718e-6,
                    'η_iiii': 4.789e-8
                },
                'spectroscopic_constants': {
                    'band_origin': {
                        'literature': 1556.39,   # Yu et al. (2014)
                        'tosh': 1556.43,
                        'harmonic': 1580.00,
                        'error_tosh': 0.04,
                        'error_harmonic': 23.61
                    },
                    'rotational_constants': {
                        'B0': {'literature': 1.4377, 'harmonic': 1.4456, 'tosh': 1.4257},
                        'B1': {'literature': 1.4219, 'harmonic': 1.4456, 'tosh': 1.4257}
                    },
                    'reduced_mass': 8.0,
                    'equilibrium_length': 1.207e-8,
                    'force_constant': 1.177e3,
                    'branches_present': ['O', 'Q', 'S'],
                    'max_quantum_number': 50
                },
                'references': [
                    "Yu et al. (2014)",
                    "Domagal-Goldman et al. (2014)",
                    "Meadows (2017)"
                ]
            },

            'N2': {
                'name': 'Diatomic Nitrogen',
                'formula': 'N₂',
                'molecular_weight': 28.014,
                'type': 'diatomic',
                'rotor_type': 'linear',
                'symmetry': 'homonuclear',
                'astrophysical_context': (
                    "Key species in cosmology, observed in various galactic environments. "
                    "Direct observation difficult due to lack of strong pure rotational "
                    "or vibrational lines."
                ),
                'biogenic_context': (
                    "Essential ingredient for building blocks of life. Required along with "
                    "carbon and phosphorus for formation of nucleic acids and proteins."
                ),
                'solar_system_presence': {
                    'Venus': '3.5%',
                    'Earth': '78.1%',
                    'Mars': '2.8%',
                    'Titan': '98.4%'
                },
                'detection_methods': [
                    "Indirect methods",
                    "Weak quadrupole transitions"
                ],
                'spectral_activity': 'Raman active, IR inactive',
                'nuclear_spin': 'Non-zero (alternating intensities)',
                'anharmonic_constants': {
                    'η_ii': 1.156e-4,
                    'η_iii': -4.048e-6,
                    'η_iiii': 1.069e-7
                },
                'spectroscopic_constants': {
                    'band_origin': {
                        'literature': 2329.91,   # Bendtsen and Rasmussen (2000)
                        'tosh': 2329.85,
                        'harmonic': 2359.56,
                        'error_tosh': 0.06,
                        'error_harmonic': 29.65
                    },
                    'rotational_constants': {
                        'B0': {'literature': 1.9896, 'harmonic': 1.9982, 'tosh': 1.9753},
                        'B1': {'literature': 1.9722, 'harmonic': 1.9982, 'tosh': 1.9753}
                    },
                    'reduced_mass': 7.0,
                    'equilibrium_length': 1.098e-8,
                    'force_constant': 2.298e3,
                    'branches_present': ['O', 'Q', 'S'],
                    'max_quantum_number': 40
                },
                'references': [
                    "Bendtsen and Rasmussen (2000)",
                    "Edwards et al. (1993)",
                    "Lammer et al. (2019)"
                ]
            },

            'CO': {
                'name': 'Carbon Monoxide',
                'formula': 'CO',
                'molecular_weight': 28.01,
                'type': 'diatomic',
                'rotor_type': 'linear',
                'symmetry': 'heteronuclear',
                'astrophysical_context': (
                    "Has permanent dipole moment, readily absorbs even at low temperatures "
                    "found in molecular clouds. Already detected in exoplanet atmosphere "
                    "(HD 189733b)."
                ),
                'biogenic_context': (
                    "Classified as anti-biosignature. Difficult to accumulate in atmosphere "
                    "of inhabited planets as it acts as energy source for some microbes. "
                    "Can indicate technosignature if from incomplete combustion."
                ),
                'detection_methods': [
                    "Direct IR spectroscopy",
                    "Rotational-vibrational transitions"
                ],
                'spectral_activity': 'IR active, Raman inactive',
                'nuclear_spin': 'Mixed',
                'anharmonic_constants': {
                    'η_ii': 9.803e-5,
                    'η_iii': -3.430e-6,
                    'η_iiii': 8.928e-8
                },
                'spectroscopic_constants': {
                    'band_origin': {
                        'literature': 2143.27,   # Coxon and Hajigeorgiou (2004)
                        'tosh': 2143.15,
                        'harmonic': 2173.07,
                        'error_tosh': 0.12,
                        'error_harmonic': 29.80
                    },
                    'rotational_constants': {
                        'B0': {'literature': 1.9225, 'harmonic': 1.9316, 'tosh': 1.9080},
                        'B1': {'literature': 1.9050, 'harmonic': 1.9316, 'tosh': 1.9080}
                    },
                    'reduced_mass': 6.856,
                    'equilibrium_length': 1.128e-8,
                    'force_constant': 1.902e3,
                    'branches_present': ['P', 'R'],
                    'max_quantum_number': 50
                },
                'references': [
                    "Coxon and Hajigeorgiou (2004)",
                    "Meshkov et al. (2018)",
                    "de Kok et al. (2013)"
                ]
            }
        }

        # TRIATOMIC DATA with asymmetric rotor capabilities
        self.triatomic_molecules = {
            'H2O': {
                'name': 'Water',
                'formula': 'H₂O',
                'molecular_weight': 18.015,
                'type': 'triatomic',
                'rotor_type': 'asymmetric',
                'symmetry': 'C2v',
                'linear': False,
                'astrophysical_context': (
                    "Universal solvent, essential for life as we know it. "
                    "Strong biosignature, but also abundant from abiotic processes."
                ),
                'biogenic_context': (
                    "Essential for all known life forms. Liquid water is considered "
                    "a prerequisite for life. Water ice and vapor can indicate "
                    "habitable environments."
                ),
                'rotational_constants_cm': {
                    'ground': {'A': 27.88, 'B': 14.52, 'C': 9.28}
                },
                'epimetheus_results': {  # TOSH calculations
                    'fundamental_frequencies': {
                        'ν1': 4067.9,  # cm⁻¹ (symmetric stretch)
                        'ν2': 2112.5,  # cm⁻¹ (bending)
                        'ν3': 4255.3   # cm⁻¹ (asymmetric stretch)
                    },
                    'ground_state_rotational_constants': {
                        'A': 23.29,    # cm⁻¹
                        'B': 14.55,    # cm⁻¹
                        'C': 8.96      # cm⁻¹
                    }
                },
                'vpt2_results': {  # VPT2 calculations
                    'fundamental_frequencies': {
                        'ν1': 4014.9,  # cm⁻¹
                        'ν2': 2123.7,  # cm⁻¹
                        'ν3': 4265.5   # cm⁻¹
                    },
                    'ground_state_rotational_constants': {
                        'A': 23.57,    # cm⁻¹
                        'B': 14.51,    # cm⁻¹
                        'C': 8.79      # cm⁻¹
                    }
                },
                'molecular_properties': {
                    'geometry': 'Bent (C2v symmetry)',
                    'bond_angle': 104.51,
                    'bond_length': 0.9576257,  # Å
                    'dipole_moment': 1.85,     # Debye
                    'point_group': 'C2v'
                },
                'detection_methods': [
                    "IR spectroscopy (multiple bands)",
                    "Submillimeter observations",
                    "Transmission spectroscopy of exoplanet atmospheres"
                ]
            },

            'O3': {
                'name': 'Ozone',
                'formula': 'O₃',
                'molecular_weight': 47.998,
                'type': 'triatomic',
                'rotor_type': 'asymmetric',
                'symmetry': 'C2v',
                'linear': False,
                'astrophysical_context': (
                    "Strong UV absorber, crucial for Earth's ozone layer. "
                    "Can be a biosignature for oxygen-rich atmospheres."
                ),
                'biogenic_context': (
                    "On Earth, ozone is primarily produced photochemically from "
                    "oxygen produced by photosynthesis. Can indicate biological "
                    "oxygen production."
                ),
                'rotational_constants_cm': {
                    'ground': {'A': 3.429, 'B': 0.443, 'C': 0.392}
                },
                'epimetheus_results': {
                    'fundamental_frequencies': {
                        'ν1': 1531.5,  # cm⁻¹ (symmetric stretch)
                        'ν2': 672.8,   # cm⁻¹ (bending)
                        'ν3': 1440.1   # cm⁻¹ (asymmetric stretch)
                    },
                    'ground_state_rotational_constants': {
                        'A': 3.429,    # cm⁻¹
                        'B': 0.443,    # cm⁻¹
                        'C': 0.392     # cm⁻¹
                    }
                },
                'molecular_properties': {
                    'geometry': 'Bent (C2v symmetry)',
                    'bond_angle': 116.8,
                    'bond_length': 1.278e-8,  # cm
                    'dipole_moment': 0.53  # Debye
                },
                'detection_methods': [
                    "UV absorption spectroscopy",
                    "IR spectroscopy (9.6 μm band)",
                    "Microwave spectroscopy"
                ]
            }
        }

        # Additional asymmetric tops
        self.polyatomic_molecules = {
            'NH3': {
                'name': 'Ammonia',
                'formula': 'NH₃',
                'type': 'polyatomic',
                'rotor_type': 'symmetric_oblate',
                'symmetry': 'C3v',
                'rotational_constants_cm': {
                    'ground': {'A': 9.94, 'B': 9.94, 'C': 6.30}
                },
                'astrophysical_context': (
                    "Common in cold interstellar clouds. Can be a biosignature "
                    "when combined with methane in reducing atmospheres."
                ),
                'biogenic_context': (
                    "Produced by anaerobic decomposition of organic matter. "
                    "Present in Earth's atmosphere at trace levels."
                )
            },

            'CH4': {
                'name': 'Methane',
                'formula': 'CH₄',
                'type': 'polyatomic',
                'rotor_type': 'spherical',
                'symmetry': 'Td',
                'rotational_constants_cm': {
                    'ground': {'A': 5.24, 'B': 5.24, 'C': 5.24}
                },
                'astrophysical_context': (
                    "Major component of gas giants' atmospheres. "
                    "Potential biosignature when found with oxygen."
                ),
                'biogenic_context': (
                    "On Earth, primarily produced by methanogenic archaea. "
                    "Also from geothermal and volcanic activity."
                )
            },

            'NO2': {
                'name': 'Nitrogen Dioxide',
                'formula': 'NO₂',
                'type': 'triatomic',
                'rotor_type': 'asymmetric',
                'symmetry': 'C2v',
                'linear': False,
                'rotational_constants_cm': {
                    'ground': {'A': 8.00, 'B': 0.43, 'C': 0.41}
                },
                'astrophysical_context': (
                    "Potential indicator of atmospheric photochemistry. "
                    "Could indicate abiotic nitrogen fixation."
                )
            }
        }

        # Merge all molecules
        self.molecules = {}
        self.molecules.update(self.diatomic_molecules)
        self.molecules.update(self.triatomic_molecules)
        self.molecules.update(self.polyatomic_molecules)

        # Theoretical methods
        self.theoretical_methods = {
            'TOSH': {
                'full_name': 'Taylor-series-based One-mode Shifted Harmonic',
                'description': 'Anharmonic correction method using σ shift parameter',
                'advantages': [
                    'Lower computational cost than VPT2',
                    'No need for η_ijk derivatives',
                    'Good accuracy for most vibrational modes'
                ]
            },

            'VPT2': {
                'full_name': 'Vibrational Perturbation Theory to Second Order',
                'description': 'Full second-order perturbation theory for anharmonic corrections',
                'advantages': [
                    'Higher accuracy than TOSH',
                    'Includes all anharmonic terms',
                    'Better for symmetric stretches'
                ]
            },

            'Harmonic': {
                'full_name': 'Harmonic Oscillator Approximation',
                'description': 'Simple harmonic oscillator model',
                'advantages': [
                    'Extremely low computational cost',
                    'Analytic solutions available',
                    'Good for initial estimates'
                ]
            }
        }

        # Calculate asymmetry parameters for all molecules
        self._calculate_asymmetry_parameters()

        print(f"✓ Database initialized with:")
        print(f"  • {len(self.diatomic_molecules)} diatomic molecules (real data)")
        print(f"  • {len(self.triatomic_molecules)} triatomic molecules")
        print(f"  • {len(self.polyatomic_molecules)} polyatomic molecules")
        print(f"  • {len(self.theoretical_methods)} theoretical methods")

    def _calculate_asymmetry_parameters(self):
        """Calculate Ray's asymmetry parameter for all molecules"""
        for name, mol in self.molecules.items():
            if 'rotational_constants_cm' in mol and 'ground' in mol['rotational_constants_cm']:
                consts = mol['rotational_constants_cm']['ground']
                if all(k in consts for k in ['A', 'B', 'C']):
                    A, B, C = consts['A'], consts['B'], consts['C']
                    # Skip linear molecules where A=0 or B≈C
                    if A == 0 and abs(B - C) < 0.1:
                        continue
                    try:
                        kappa = self.asymmetric_theory.calculate_ray_asymmetry_parameter(A, B, C)
                        if 'molecular_properties' not in mol:
                            mol['molecular_properties'] = {}
                        mol['molecular_properties']['ray_asymmetry'] = kappa
                        mol['molecular_properties']['asymmetry_classification'] = \
                            self.asymmetric_theory.classify_asymmetric_top(kappa)
                    except Exception as e:
                        print(f"  ⚠ Could not calculate κ for {name}: {str(e)[:50]}")

    def get_molecule(self, molecule_code: str) -> Dict:
        """Get complete molecular data"""
        molecule_code = molecule_code.upper()

        if molecule_code in self.molecules:
            mol = self.molecules[molecule_code].copy()
            mol['code'] = molecule_code
            return mol

        # Check variations
        variations = {
            'H2O': 'H2O', 'WATER': 'H2O', 'H₂O': 'H2O',
            'O3': 'O3', 'OZONE': 'O3', 'O₃': 'O3',
            'CO2': 'CO2', 'CARBONDIOXIDE': 'CO2',
            'CH4': 'CH4', 'METHANE': 'CH4',
            'NH3': 'NH3', 'AMMONIA': 'NH3',
            'NO2': 'NO2', 'NITROGEN_DIOXIDE': 'NO2',
            'H2': 'H2', 'HYDROGEN': 'H2', 'H₂': 'H2',
            'O2': 'O2', 'OXYGEN': 'O2', 'O₂': 'O2',
            'N2': 'N2', 'NITROGEN': 'N2', 'N₂': 'N2',
            'CO': 'CO', 'CARBON_MONOXIDE': 'CO'
        }

        if molecule_code in variations:
            return self.get_molecule(variations[molecule_code])

        return {}

    def get_spectroscopic_summary(self, molecule_code: str) -> pd.DataFrame:
        """Get spectroscopic data as DataFrame (from real data)"""
        mol = self.get_molecule(molecule_code)
        if not mol:
            return pd.DataFrame()

        # Check if this is a diatomic with real data structure
        if molecule_code in self.diatomic_molecules:
            data = {
                'Parameter': [],
                'Literature Value': [],
                'TOSH Value': [],
                'Harmonic Value': [],
                'TOSH Error': [],
                'Harmonic Error': []
            }

            spec = mol['spectroscopic_constants']

            # Band origin
            data['Parameter'].append('Band Origin (cm⁻¹)')
            data['Literature Value'].append(spec['band_origin']['literature'])
            data['TOSH Value'].append(spec['band_origin']['tosh'])
            data['Harmonic Value'].append(spec['band_origin']['harmonic'])
            data['TOSH Error'].append(spec['band_origin']['error_tosh'])
            data['Harmonic Error'].append(spec['band_origin']['error_harmonic'])

            # B0
            data['Parameter'].append('B₀ (cm⁻¹)')
            data['Literature Value'].append(spec['rotational_constants']['B0']['literature'])
            data['TOSH Value'].append(spec['rotational_constants']['B0']['tosh'])
            data['Harmonic Value'].append(spec['rotational_constants']['B0']['harmonic'])
            data['TOSH Error'].append(abs(spec['rotational_constants']['B0']['literature'] -
                                          spec['rotational_constants']['B0']['tosh']))
            data['Harmonic Error'].append(abs(spec['rotational_constants']['B0']['literature'] -
                                             spec['rotational_constants']['B0']['harmonic']))

            # B1
            if 'B1' in spec['rotational_constants']:
                data['Parameter'].append('B₁ (cm⁻¹)')
                data['Literature Value'].append(spec['rotational_constants']['B1']['literature'])
                data['TOSH Value'].append(spec['rotational_constants']['B1']['tosh'])
                data['Harmonic Value'].append(spec['rotational_constants']['B1']['harmonic'])
                data['TOSH Error'].append(abs(spec['rotational_constants']['B1']['literature'] -
                                              spec['rotational_constants']['B1']['tosh']))
                data['Harmonic Error'].append(abs(spec['rotational_constants']['B1']['literature'] -
                                                 spec['rotational_constants']['B1']['harmonic']))

            df = pd.DataFrame(data)
            df['TOSH Improvement'] = df['Harmonic Error'] - df['TOSH Error']

            return df
        else:
            # For non-diatomic molecules, create a simpler summary
            data = []
            if 'rotational_constants_cm' in mol:
                consts = mol['rotational_constants_cm'].get('ground', {})
                for const_name, value in consts.items():
                    data.append({
                        'Parameter': f'{const_name} (cm⁻¹)',
                        'Value': value,
                        'Type': 'Rotational Constant'
                    })

            if 'epimetheus_results' in mol:
                freqs = mol['epimetheus_results'].get('fundamental_frequencies', {})
                for freq_name, value in freqs.items():
                    data.append({
                        'Parameter': f'{freq_name} (cm⁻¹)',
                        'Value': value,
                        'Type': 'Fundamental Frequency (TOSH)'
                    })

            return pd.DataFrame(data)

    def compare_methods(self) -> pd.DataFrame:
        """Compare TOSH vs Harmonic accuracy for all molecules (real data only)"""
        data = []

        for code, mol in self.diatomic_molecules.items():  # Only diatomic have real data comparison
            spec = mol['spectroscopic_constants']

            # Calculate average errors
            tosh_error_band = spec['band_origin']['error_tosh']
            harmonic_error_band = spec['band_origin']['error_harmonic']

            tosh_error_b0 = abs(spec['rotational_constants']['B0']['literature'] -
                               spec['rotational_constants']['B0']['tosh'])
            harmonic_error_b0 = abs(spec['rotational_constants']['B0']['literature'] -
                                   spec['rotational_constants']['B0']['harmonic'])

            tosh_error_b1 = 0
            harmonic_error_b1 = 0
            if 'B1' in spec['rotational_constants']:
                tosh_error_b1 = abs(spec['rotational_constants']['B1']['literature'] -
                                   spec['rotational_constants']['B1']['tosh'])
                harmonic_error_b1 = abs(spec['rotational_constants']['B1']['literature'] -
                                       spec['rotational_constants']['B1']['harmonic'])

            avg_tosh_error = np.mean([tosh_error_band, tosh_error_b0, tosh_error_b1])
            avg_harmonic_error = np.mean([harmonic_error_band, harmonic_error_b0, harmonic_error_b1])

            improvement = ((harmonic_error_band - tosh_error_band) / harmonic_error_band * 100
                          if harmonic_error_band != 0 else 0)

            data.append({
                'Molecule': mol['name'],
                'Formula': mol['formula'],
                'TOSH Band Error (cm⁻¹)': tosh_error_band,
                'Harmonic Band Error (cm⁻¹)': harmonic_error_band,
                'Band Improvement (%)': improvement,
                'TOSH B₀ Error (cm⁻¹)': tosh_error_b0,
                'Harmonic B₀ Error (cm⁻¹)': harmonic_error_b0,
                'TOSH B₁ Error (cm⁻¹)': tosh_error_b1,
                'Harmonic B₁ Error (cm⁻¹)': harmonic_error_b1,
                'Avg TOSH Error': avg_tosh_error,
                'Avg Harmonic Error': avg_harmonic_error,
                'Method Preference': 'TOSH' if avg_tosh_error < avg_harmonic_error else 'Harmonic'
            })

        return pd.DataFrame(data)

    def get_polyatomic_classification_summary(self) -> pd.DataFrame:
        """Get summary of polyatomic molecule classifications"""
        data = []

        for code, mol in self.molecules.items():
            if mol['type'] != 'diatomic':  # Focus on polyatomic
                ray_asymmetry = mol.get('molecular_properties', {}).get('ray_asymmetry', 'N/A')
                if ray_asymmetry != 'N/A':
                    ray_asymmetry = f"{ray_asymmetry:.3f}"

                data.append({
                    'Molecule': mol['name'],
                    'Formula': mol['formula'],
                    'Rotor Type': mol.get('rotor_type', 'unknown'),
                    'Symmetry': mol.get('symmetry', 'unknown'),
                    'Asymmetry κ': ray_asymmetry,
                    'Classification': mol.get('molecular_properties', {}).get('asymmetry_classification', 'N/A')
                })

        return pd.DataFrame(data)

    def calculate_anharmonic_corrections(self, molecule_code: str) -> Dict:
        """Calculate anharmonic corrections from constants (for diatomic molecules)"""
        mol = self.get_molecule(molecule_code)
        if not mol or molecule_code not in self.diatomic_molecules:
            return {}

        const = mol['anharmonic_constants']

        # Convert to SI units for calculations (simplified)
        η_ii_si = const['η_ii'] * 4.3597447222071e-18  # Hartree to Joule

        results = {
            'anharmonicity_constant': const['η_ii'],
            'cubic_term': const['η_iii'],
            'quartic_term': const['η_iiii'],
            'relative_strength': abs(const['η_iii']/const['η_ii']) if const['η_ii'] != 0 else 0,
            'anharmonicity_type': 'Strong' if abs(const['η_iii']/const['η_ii']) > 0.1 else 'Moderate'
        }

        return results

# Initialize database
comp_db = ComprehensiveMolecularDatabase()

# ============================================================================
# SECTION 5: TOSH THEORY IMPLEMENTATION (ENHANCED WITH REAL DATA)
# ============================================================================
print("\n" + "="*70)
print("TOSH THEORY IMPLEMENTATION WITH REAL MOLECULAR DATA")
print("="*70)

class TOSHCalculatorEnhanced:
    """Enhanced TOSH calculator with real molecular data"""

    def __init__(self, molecule_code='H2'):
        """
        Initialize TOSH calculator with real molecular data

        Parameters:
        molecule_code: Molecular code ('H2', 'O2', 'N2', 'CO', etc.)
        """
        self.molecule_code = molecule_code.upper()
        self.molecular_data = comp_db.get_molecule(molecule_code)

        if not self.molecular_data:
            raise ValueError(f"Molecule {molecule_code} not found in database")

        # Check if this is a diatomic molecule with real data
        if molecule_code in comp_db.diatomic_molecules:
            # Extract constants from real data
            const = self.molecular_data['anharmonic_constants']
            spec = self.molecular_data['spectroscopic_constants']

            self.eta_ii = const['η_ii']
            self.eta_iii = const['η_iii']
            self.eta_iiii = const['η_iiii']

            # Estimate omega from band origin (simplified)
            self.omega_est = spec['band_origin']['harmonic']

            # Reduced mass
            self.reduced_mass = spec['reduced_mass']

            # Calculate shift parameter σ (Eq. 2.65 from thesis)
            self.sigma = -5/12 * self.eta_iii / self.omega_est**3

            print(f"✓ TOSH calculator initialized for {self.molecular_data['name']} (diatomic)")
            print(f"  σ = {self.sigma:.6e}")
        else:
            # For non-diatomic molecules, use simplified approach
            print(f"⚠ {molecule_code} is not a diatomic molecule. Using simplified TOSH approach.")
            self.eta_ii = 1e-4  # Default values
            self.eta_iii = -1e-5
            self.eta_iiii = 1e-6
            self.omega_est = 1000  # Default frequency
            self.reduced_mass = 10  # Default mass
            self.sigma = 0.01  # Default shift

    def ground_state_energy(self) -> float:
        """Calculate ground state energy E₀ (Eq. 2.58)"""
        E0 = (1/4) * self.omega_est
        E0 += (1/2) * self.eta_ii * (1/(2*self.omega_est) + self.sigma**2)
        E0 += (1/6) * self.eta_iii * (3*self.sigma/(2*self.omega_est) + self.sigma**3)
        E0 += (1/24) * self.eta_iiii * (3/(4*self.omega_est**2) +
                                       3*self.sigma**2/self.omega_est + self.sigma**4)
        return E0

    def excited_state_energy(self) -> float:
        """Calculate first excited state energy E₁ (Eq. 2.60)"""
        E1 = (3/4) * self.omega_est
        E1 += (1/2) * self.eta_ii * (3/(2*self.omega_est) + self.sigma**2)
        E1 += (1/6) * self.eta_iii * (9*self.sigma/(2*self.omega_est) + self.sigma**3)
        E1 += (1/24) * self.eta_iiii * (15/(4*self.omega_est**2) +
                                       9*self.sigma**2/self.omega_est + self.sigma**4)
        return E1

    def fundamental_transition(self) -> Dict:
        """Calculate fundamental transition energy ΔE^TOSH (Eq. 2.62)"""
        E0 = self.ground_state_energy()
        E1 = self.excited_state_energy()
        delta_E = E1 - E0

        # Check if we have literature value for comparison
        if self.molecule_code in comp_db.diatomic_molecules:
            lit_value = self.molecular_data['spectroscopic_constants']['band_origin']['literature']
            tosh_value = self.molecular_data['spectroscopic_constants']['band_origin']['tosh']
        else:
            lit_value = delta_E * 0.95  # Simulated literature value
            tosh_value = delta_E

        error = abs(delta_E - lit_value)
        relative_error = (error / lit_value * 100) if lit_value != 0 else 0

        return {
            'calculated': delta_E,
            'literature': lit_value,
            'tosh_database': tosh_value if 'tosh_value' in locals() else delta_E,
            'absolute_error': error,
            'relative_error': relative_error,
            'method_accuracy': 'Excellent' if error < 1 else 'Good' if error < 10 else 'Fair'
        }

class EnhancedQuantumHarmonicOscillator:
    """Quantum Harmonic Oscillator with real molecular data"""

    def __init__(self, molecule_code='H2', use_real_data=True):
        """
        Initialize QHO with real molecular data

        Parameters:
        molecule_code: 'H2', 'O2', 'N2', 'CO', 'H2O', etc.
        use_real_data: Whether to use real constants from database
        """
        self.molecule_code = molecule_code.upper()
        self.use_real_data = use_real_data

        if use_real_data:
            mol_data = comp_db.get_molecule(molecule_code)
            if not mol_data:
                print(f"Warning: {molecule_code} not found in database. Using defaults.")
                mol_data = comp_db.get_molecule('H2')

            self.molecular_data = mol_data

            # Get spectroscopic data based on molecule type
            if molecule_code in comp_db.diatomic_molecules:
                spec = mol_data['spectroscopic_constants']

                # Convert to SI units
                self.reduced_mass = spec['reduced_mass'] * 1.66053906660e-27  # amu to kg
                self.force_constant = spec['force_constant']  # Already in N/m

                # Store spectroscopic data
                self.band_origin = {
                    'literature': spec['band_origin']['literature'],
                    'tosh': spec['band_origin']['tosh'],
                    'harmonic': spec['band_origin']['harmonic']
                }

                if 'rotational_constants' in spec:
                    self.rotational_constants = {
                        'B0': spec['rotational_constants']['B0'],
                        'B1': spec['rotational_constants']['B1']
                    }
            else:
                # For non-diatomic, use defaults
                self.reduced_mass = 10 * 1.66053906660e-27
                self.force_constant = 1.0e3
                self.band_origin = {
                    'literature': 1000,
                    'tosh': 1000,
                    'harmonic': 1000
                }
                self.rotational_constants = {
                    'B0': {'literature': 1.0, 'tosh': 1.0, 'harmonic': 1.0},
                    'B1': {'literature': 1.0, 'tosh': 1.0, 'harmonic': 1.0}
                }
        else:
            # Default values
            self.reduced_mass = 1.66053906660e-27  # 1 amu in kg
            self.force_constant = 1.0
            self.molecular_data = {}

        # Physical constants
        self.hbar = 1.054571817e-34  # Reduced Planck constant (J·s)
        self.h = 6.62607015e-34      # Planck constant (J·s)
        self.c = 299792458           # Speed of light (m/s)
        self.k_B = 1.380649e-23      # Boltzmann constant (J/K)

        # Calculate derived parameters
        self.omega = np.sqrt(self.force_constant / self.reduced_mass)  # Angular frequency (rad/s)
        self.frequency = self.omega / (2 * np.pi)                      # Frequency (Hz)
        self.wavenumber = self.frequency / (self.c * 100)              # Convert to cm⁻¹

        # Anharmonic constants if available
        if use_real_data and 'anharmonic_constants' in self.molecular_data:
            self.anharmonic_constants = self.molecular_data['anharmonic_constants']
        else:
            self.anharmonic_constants = None

    def energy_level(self, n: int, include_anharmonic=False) -> float:
        """
        Calculate energy for quantum number n

        Parameters:
        n: Quantum number
        include_anharmonic: Whether to include anharmonic corrections

        Returns:
        Energy in Joules
        """
        # Harmonic oscillator energy
        E_harmonic = self.hbar * self.omega * (n + 0.5)

        if include_anharmonic and self.anharmonic_constants:
            # Simplified anharmonic correction
            anharmonicity = 0.01  # Approximate anharmonicity constant
            E_anharmonic = E_harmonic - anharmonicity * E_harmonic**2 / (self.hbar * self.omega)
            return E_anharmonic

        return E_harmonic

    def thermal_population(self, n: int, temperature: float = 300.0) -> float:
        """
        Calculate Boltzmann population at temperature T

        Parameters:
        n: Quantum number
        temperature: Temperature in Kelvin

        Returns:
        Population fraction
        """
        # Calculate partition function (sum over first 20 levels)
        partition = 0.0
        energies = []

        for i in range(20):
            E_i = self.energy_level(i)
            energies.append(E_i)
            partition += np.exp(-E_i / (self.k_B * temperature))

        # Population of level n
        E_n = self.energy_level(n)
        population = np.exp(-E_n / (self.k_B * temperature)) / partition

        return population

# ============================================================================
# SECTION 6: PANDORA SPECTRAL MODELING (WITH ASYMMETRIC ROTOR SUPPORT)
# ============================================================================
print("\n" + "="*70)
print("PANDORA SPECTRAL MODELING WITH ASYMMETRIC ROTOR SUPPORT")
print("="*70)

class PandoraSpectralModeler:
    """Implementation of Pandora spectral modeling code with asymmetric rotor support"""

    def __init__(self, molecule_code: str = 'H2O'):
        self.molecule_code = molecule_code.upper()
        self.molecule = comp_db.get_molecule(molecule_code)

        if not self.molecule:
            raise ValueError(f"Molecule {molecule_code} not found")

        self.asymmetric_theory = AsymmetricRotorTheory()

        print(f"✓ Pandora initialized for {self.molecule['name']}")
        print(f"  • Rotor type: {self.molecule.get('rotor_type', 'unknown')}")
        print(f"  • Type: {self.molecule.get('type', 'unknown')}")

    def simulate_asymmetric_rotor_spectrum(self, temperature: float = 300.0,
                                          J_max: int = 10,
                                          band_type: str = 'a') -> Dict:
        """Simulate asymmetric rotor spectrum"""
        # Get rotational constants
        if 'rotational_constants_cm' not in self.molecule:
            # Try to get from diatomic data
            if self.molecule_code in comp_db.diatomic_molecules:
                spec = self.molecule['spectroscopic_constants']
                A = 0  # Linear molecule, A=0
                B = spec['rotational_constants']['B0']['literature']
                C = B  # Linear molecule, B=C
            else:
                raise ValueError(f"No rotational constants for {self.molecule_code}")
        else:
            consts = self.molecule['rotational_constants_cm'].get('ground', {})
            if not all(k in consts for k in ['A', 'B', 'C']):
                raise ValueError(f"Incomplete rotational constants")

            A, B, C = consts['A'], consts['B'], consts['C']

        # Calculate asymmetry parameter
        kappa = self.asymmetric_theory.calculate_ray_asymmetry_parameter(A, B, C)

        # Generate simple transitions
        transitions = self._generate_simple_transitions(A, B, C, temperature, J_max, band_type)

        return {
            'molecule': self.molecule['name'],
            'formula': self.molecule['formula'],
            'temperature': temperature,
            'band_type': band_type,
            'rotational_constants': {'A': A, 'B': B, 'C': C},
            'ray_asymmetry': kappa,
            'asymmetry_classification': self.asymmetric_theory.classify_asymmetric_top(kappa),
            'transitions': transitions
        }

    def _generate_simple_transitions(self, A: float, B: float, C: float,
                                    temperature: float, J_max: int,
                                    band_type: str) -> Dict:
        """Generate simple transitions for demonstration"""
        # Use effective B constant
        B_eff = (A + B + C) / 3

        transitions = {
            'frequencies': [],
            'intensities': [],
            'J_lower': [],
            'J_upper': []
        }

        # Generate P, Q, R branches
        for J_lower in range(J_max + 1):
            for delta_J in [-1, 0, 1]:  # P, Q, R branches
                J_upper = J_lower + delta_J

                if J_upper < 0 or J_upper > J_max:
                    continue

                # Simplified frequency calculation
                frequency = abs(delta_J) * B_eff * 2  # Simplified

                # Boltzmann intensity
                E_lower = B_eff * J_lower * (J_lower + 1)
                E_lower_J = E_lower * self.asymmetric_theory.h * self.asymmetric_theory.c * 100
                boltzmann = np.exp(-E_lower_J / (self.asymmetric_theory.k_B * temperature))
                intensity = (2 * J_lower + 1) * boltzmann

                transitions['frequencies'].append(frequency)
                transitions['intensities'].append(intensity)
                transitions['J_lower'].append(J_lower)
                transitions['J_upper'].append(J_upper)

        # Normalize intensities
        if transitions['intensities']:
            max_intensity = max(transitions['intensities'])
            transitions['intensities'] = [i/max_intensity for i in transitions['intensities']]

        return transitions

    def calculate_spectrum(self, temperature: float = 300.0,
                          j_max: int = 15) -> Dict:
        """Calculate complete rovibrational spectrum"""
        # Get fundamental transition using TOSH
        try:
            tosh_calc = TOSHCalculatorEnhanced(self.molecule_code)
            fundamental = tosh_calc.fundamental_transition()
        except:
            # Fallback if TOSH fails
            fundamental = {
                'calculated': 1000,  # Default
                'literature': 1000,
                'tosh_database': 1000,
                'absolute_error': 0,
                'relative_error': 0,
                'method_accuracy': 'Unknown'
            }

        # Get rotational constants
        if self.molecule_code in comp_db.diatomic_molecules:
            spec = self.molecule['spectroscopic_constants']
            B0 = spec['rotational_constants']['B0']['literature']
            B1 = spec['rotational_constants']['B1']['literature']
            branch_types = spec.get('branches_present', ['P', 'Q', 'R'])
        else:
            # For non-diatomic, use asymmetric rotor simulation
            spectrum = self.simulate_asymmetric_rotor_spectrum(temperature, j_max, 'a')
            return spectrum

        nu_0 = fundamental['calculated']

        # Initialize results
        spectrum = {
            'molecule': self.molecule['name'],
            'formula': self.molecule['formula'],
            'band_origin': fundamental,
            'temperature': temperature,
            'j_max': j_max,
            'branches': {}
        }

        # Calculate each branch
        for branch in branch_types:
            transitions = []

            for J in range(j_max + 1):
                if branch == 'P' and J > 0:  # ΔJ = -1
                    freq = nu_0 - (B1 + B0) * J + (B1 - B0) * J**2
                    degeneracy = 2 * J + 1

                elif branch == 'Q':  # ΔJ = 0
                    freq = nu_0 + (B1 - B0) * J * (J + 1)
                    degeneracy = 2 * J + 1

                elif branch == 'R':  # ΔJ = +1
                    freq = nu_0 + (B1 + B0) * (J + 1) + (B1 - B0) * (J + 1)**2
                    degeneracy = 2 * J + 1

                else:
                    continue

                # Calculate Boltzmann factor
                E_rot = B0 * J * (J + 1) * 1.98630e-23  # Convert to Joules
                boltzmann = np.exp(-E_rot / (1.380649e-23 * temperature))
                intensity = degeneracy * boltzmann

                transitions.append({
                    'J': J,
                    'frequency': freq,
                    'intensity': intensity,
                    'degeneracy': degeneracy
                })

            if transitions:
                # Normalize intensities
                max_intensity = max(t['intensity'] for t in transitions)
                for t in transitions:
                    t['relative_intensity'] = t['intensity'] / max_intensity

                spectrum['branches'][branch] = transitions

        return spectrum

# ============================================================================
# SECTION 7: EXOPLANET DISCOVERY DATA INTEGRATION
# ============================================================================
print("\n" + "="*70)
print("EXOPLANET DISCOVERY DATA INTEGRATION")
print("="*70)

class ExoplanetDiscoveryData:
    """Class to handle exoplanet discovery data with NASA data compatibility"""

    def __init__(self, use_sample_data=True):
        """
        Initialize exoplanet data handler

        Parameters:
        use_sample_data: If True, use enhanced sample data instead of NASA data
        """
        self.exoplanet_data = None
        self.is_nasa_data = False
        self.use_sample_data = use_sample_data
        self.load_exoplanet_data()

    def load_exoplanet_data(self):
        """Load exoplanet data from NASA Exoplanet Archive or create sample data"""
        if self.use_sample_data:
            print("Using enhanced sample exoplanet data with theoretical features...")
            self.create_enhanced_sample_data()
            self.is_nasa_data = False
        else:
            try:
                # Try to load from NASA Exoplanet Archive
                print("Attempting to load data from NASA Exoplanet Archive...")
                url = "https://exoplanetarchive.ipac.caltech.edu/TAP/sync?query=select+*+from+ps&format=csv"
                self.exoplanet_data = pd.read_csv(url)
                self.is_nasa_data = True
                print(f"✓ Loaded {len(self.exoplanet_data)} exoplanets from NASA Exoplanet Archive")
                print("  Note: NASA data doesn't include atmospheric abundances by default.")
                print("  Atmospheric data will be simulated for analysis.")

            except Exception as e:
                print(f"Online load failed: {e}")
                print("Creating enhanced sample exoplanet data with theoretical features...")
                self.create_enhanced_sample_data()
                self.is_nasa_data = False

    def create_enhanced_sample_data(self):
        """Create enhanced sample exoplanet data with theoretical features"""
        np.random.seed(42)
        n_planets = 5000

        # Base parameters
        data = {
            'pl_name': [f'Planet_{i:04d}' for i in range(n_planets)],
            'pl_orbper': np.random.exponential(100, n_planets) + 1,
            'pl_bmasse': np.random.lognormal(1.5, 1.2, n_planets),
            'pl_rade': np.random.lognormal(0.8, 0.6, n_planets) + 0.5,
            'pl_orbsmax': np.random.exponential(1.5, n_planets),
            'pl_eccen': np.random.beta(1.5, 5, n_planets),
            'st_teff': np.random.normal(5500, 1500, n_planets),
            'st_rad': np.random.lognormal(0, 0.3, n_planets),
            'st_mass': np.random.normal(1.0, 0.3, n_planets),
            'discoverymethod': np.random.choice(['Transit', 'Radial Velocity', 'Microlensing',
                                               'Imaging', 'Transit Timing Variations'],
                                               n_planets, p=[0.75, 0.15, 0.04, 0.02, 0.04]),
            'sy_dist': np.random.exponential(300, n_planets)
        }

        # Add theoretical atmospheric features
        data['h2o_abundance'] = self.calculate_atmospheric_h2o(data['pl_bmasse'], data['st_teff'], data['pl_orbsmax'])
        data['co2_abundance'] = self.calculate_atmospheric_co2(data['pl_bmasse'], data['st_teff'])
        data['ch4_abundance'] = self.calculate_atmospheric_ch4(data['pl_bmasse'], data['pl_orbsmax'])
        data['o2_abundance'] = self.calculate_atmospheric_o2(data['pl_bmasse'], data['st_teff'])
        data['co_abundance'] = self.calculate_atmospheric_co(data['pl_bmasse'], data['st_teff'])

        # Add habitability metrics
        data['equilibrium_temp'] = 280 * data['st_teff']/5777 * (data['pl_orbsmax'])**(-0.5)
        data['water_liquid_probability'] = self.calculate_water_liquid_probability(data['equilibrium_temp'],
                                                                                   data['pl_bmasse'])

        # Add biosignature scores
        data['biosignature_score'] = (
            data['h2o_abundance'] * 0.3 +
            data['o2_abundance'] * 0.4 +
            data['ch4_abundance'] * 0.3
        )

        self.exoplanet_data = pd.DataFrame(data)
        print(f"✓ Created enhanced sample data for {n_planets} exoplanets")
        print("  Includes: Atmospheric abundances, habitability metrics, biosignature scores")

    def calculate_atmospheric_h2o(self, mass, teff, sma):
        """Calculate H2O abundance based on planet properties"""
        base = np.log10(mass + 1) * 0.1
        temp_factor = np.exp(-(teff - 3000)**2 / (2 * 2000**2))
        distance_factor = np.exp(-sma / 0.5)
        return base * temp_factor * distance_factor * np.random.uniform(0.8, 1.2, len(mass))

    def calculate_atmospheric_co2(self, mass, teff):
        """Calculate CO2 abundance"""
        return 0.01 * np.sqrt(mass) * (5777 / (teff + 100))

    def calculate_atmospheric_ch4(self, mass, sma):
        """Calculate CH4 abundance"""
        return 0.001 * (sma / 0.1) * np.exp(-mass/50)

    def calculate_atmospheric_o2(self, mass, teff):
        """Calculate O2 abundance (potential biosignature)"""
        o2_prob = 0.001 * (teff/5777)**2 * np.exp(-mass/10)
        return np.random.exponential(o2_prob)

    def calculate_atmospheric_co(self, mass, teff):
        """Calculate CO abundance"""
        return 0.0005 * np.sqrt(mass) * np.exp(-teff/4000)

    def calculate_water_liquid_probability(self, temp, mass):
        """Calculate probability of liquid water"""
        temp_prob = np.exp(-(temp - 280)**2 / (2 * 50**2))
        mass_prob = 1 - np.exp(-mass/5)
        return temp_prob * mass_prob

# Initialize exoplanet data
print("\nInitializing exoplanet data...")
exoplanet_data_obj = ExoplanetDiscoveryData(use_sample_data=True)
exoplanet_data = exoplanet_data_obj.exoplanet_data
print(f"✓ Exoplanet data loaded: {len(exoplanet_data)} planets")

# ============================================================================
# SECTION 8: THEORETICAL ASTROPHYSICS MODELS
# ============================================================================
print("\n" + "="*70)
print("THEORETICAL ASTROPHYSICS MODELS")
print("="*70)

class TheoreticalAstrophysicsModels:
    """Generate theoretical models for exoplanet analysis"""

    def __init__(self):
        self.models = {}
        self.generate_models()

    def generate_models(self):
        """Generate all theoretical models"""
        print("Generating theoretical astrophysics models...")

        # 1. Molecular detection probability models
        self.models['detection_probability'] = self.generate_detection_probability_models()

        # 2. Habitable zone molecular models
        self.models['habitable_zone_chemistry'] = self.generate_habitable_zone_models()

        # 3. Biosignature confidence models
        self.models['biosignature_confidence'] = self.generate_biosignature_confidence_models()

        # 4. Molecular evolution models
        self.models['molecular_evolution'] = self.generate_molecular_evolution_models()

        print("✓ Theoretical models generated")

    def generate_detection_probability_models(self):
        """Generate models for molecular detection probabilities"""
        molecules = ['H2O', 'CO2', 'CH4', 'O2', 'O3', 'CO', 'H2', 'N2']
        abundances = np.random.uniform(0.001, 0.1, len(molecules))

        # Detection probability scales with log(abundance) and spectral strength
        detection_prob = np.log10(np.array(abundances) + 1e-10)
        detection_prob = (detection_prob - detection_prob.min()) / (detection_prob.max() - detection_prob.min())

        return {
            'molecules': molecules,
            'abundances': abundances,
            'detection_probability': detection_prob
        }

    def generate_habitable_zone_models(self):
        """Generate chemical models for habitable zones"""
        # Stellar temperature range
        stellar_temp = np.linspace(3000, 7000, 50)

        # Chemical equilibrium concentrations in habitable zone
        models = {
            'stellar_temp': stellar_temp,
            'H2O_concentration': self.calculate_hz_h2o_concentration(stellar_temp),
            'O2_concentration': self.calculate_hz_o2_concentration(stellar_temp),
            'CO2_concentration': self.calculate_hz_co2_concentration(stellar_temp),
            'CH4_concentration': self.calculate_hz_ch4_concentration(stellar_temp),
            'CO_concentration': self.calculate_hz_co_concentration(stellar_temp)
        }

        return models

    def calculate_hz_h2o_concentration(self, temp):
        """Calculate H2O concentration in habitable zone"""
        return 0.1 + 0.9 * np.exp(-(temp - 5500)**2 / (2 * 1000**2))

    def calculate_hz_o2_concentration(self, temp):
        """Calculate O2 concentration in habitable zone"""
        return 0.01 + 0.1 * np.exp(-(temp - 5800)**2 / (2 * 500**2))

    def calculate_hz_co2_concentration(self, temp):
        """Calculate CO2 concentration in habitable zone"""
        return 0.001 * (5777 / temp)**2

    def calculate_hz_ch4_concentration(self, temp):
        """Calculate CH4 concentration in habitable zone"""
        return 1e-4 * np.exp(-(temp - 5200)**2 / (2 * 800**2))

    def calculate_hz_co_concentration(self, temp):
        """Calculate CO concentration in habitable zone"""
        return 5e-5 * np.exp(-(temp - 4500)**2 / (2 * 1000**2))

    def generate_biosignature_confidence_models(self):
        """Generate models for biosignature confidence scoring"""
        biosignatures = {
            'primary': ['H2O', 'O2', 'O3', 'CH4', 'N2O'],
            'secondary': ['CO2', 'CO', 'H2S', 'CH3Cl'],
            'potential': ['PH3', 'C2H6', 'HCN', 'NH3']
        }

        confidence_scores = {}
        for category, molecules in biosignatures.items():
            for molecule in molecules:
                # Confidence based on uniqueness, stability, and detectability
                if molecule in ['O2', 'O3', 'CH4', 'N2O']:
                    confidence = np.random.uniform(0.7, 0.9)
                elif molecule in ['CO2', 'CO', 'H2O']:
                    confidence = np.random.uniform(0.4, 0.6)
                else:
                    confidence = np.random.uniform(0.2, 0.4)

                confidence_scores[molecule] = {
                    'confidence': confidence,
                    'category': category,
                    'false_positive_risk': 1 - confidence
                }

        return confidence_scores

    def generate_molecular_evolution_models(self):
        """Generate models for molecular evolution in planetary atmospheres"""
        time = np.logspace(5, 9, 50)  # 100,000 to 1 billion years

        models = {
            'time_years': time,
            'abiotic_evolution': {
                'CO2': 0.1 * np.exp(-time/1e8),
                'H2O': 0.5 * (1 - np.exp(-time/1e7)),
                'CH4': 0.01 * np.exp(-time/1e6),
                'CO': 0.05 * np.exp(-time/5e7)
            },
            'biological_evolution': {
                'O2': 0.001 * np.exp(time/1e9),
                'N2O': 1e-5 * np.exp(time/2e9),
                'CH4': 0.05 * np.exp(-time/5e8),
                'H2O': 0.8 * (1 - np.exp(-time/1e7))
            }
        }

        return models

# Initialize theoretical models
theoretical_models = TheoreticalAstrophysicsModels()

# ============================================================================
# SECTION 9: ENHANCED VISUALIZATION TOOLS
# ============================================================================
print("\n" + "="*70)
print("INITIALIZING ENHANCED VISUALIZATION TOOLS")
print("="*70)

class EnhancedAstrobiologyVisualizer:
    """Visualization tools for spectroscopic and exoplanet data"""

    def __init__(self):
        self.colors = {
            # Method colors
            'harmonic': '#FF6B6B',
            'tosh': '#4ECDC4',
            'vpt2': '#45B7D1',
            'literature': '#96CEB4',

            # Molecule colors
            'h2o': '#4ECDC4',
            'o2': '#FF6B6B',
            'co2': '#95E1D3',
            'ch4': '#D8BFD8',
            'co': '#87CEEB',
            'h2': '#FFA07A',
            'n2': '#F0E68C',
            'o3': '#DA70D6',

            # Planet type colors
            'earth_like': '#FFB6C1',
            'super_earth': '#D8BFD8',
            'mini_neptune': '#87CEEB',
            'gas_giant': '#FFA07A'
        }

    def plot_method_comparison(self):
        """Compare TOSH vs Harmonic accuracy for diatomic molecules"""
        df = comp_db.compare_methods()

        if df.empty:
            print("No comparison data available (only diatomic molecules have real data)")
            return

        fig, axes = plt.subplots(2, 2, figsize=(15, 12))

        # Plot 1: Band origin errors
        ax1 = axes[0, 0]
        x = np.arange(len(df))
        width = 0.35

        ax1.bar(x - width/2, df['TOSH Band Error (cm⁻¹)'], width,
               label='TOSH', color='blue', alpha=0.7)
        ax1.bar(x + width/2, df['Harmonic Band Error (cm⁻¹)'], width,
               label='Harmonic', color='red', alpha=0.7)

        ax1.set_xlabel('Molecule')
        ax1.set_ylabel('Error (cm⁻¹)')
        ax1.set_title('Band Origin Errors')
        ax1.set_xticks(x)
        ax1.set_xticklabels(df['Formula'], rotation=45)
        ax1.legend()
        ax1.grid(True, alpha=0.3)

        # Add improvement percentages
        for i, row in df.iterrows():
            improvement = row['Band Improvement (%)']
            ax1.text(i, max(row['TOSH Band Error (cm⁻¹)'],
                          row['Harmonic Band Error (cm⁻¹)']) + 1,
                    f'{improvement:.1f}%', ha='center', fontsize=9)

        # Plot 2: Average errors
        ax2 = axes[0, 1]
        ax2.bar(x - width/2, df['Avg TOSH Error'], width, color='blue', alpha=0.7)
        ax2.bar(x + width/2, df['Avg Harmonic Error'], width, color='red', alpha=0.7)

        ax2.set_xlabel('Molecule')
        ax2.set_ylabel('Average Error (cm⁻¹)')
        ax2.set_title('Average Spectroscopic Errors')
        ax2.set_xticks(x)
        ax2.set_xticklabels(df['Formula'], rotation=45)
        ax2.grid(True, alpha=0.3)

        # Plot 3: Rotor classification
        ax3 = axes[1, 0]
        classification_df = comp_db.get_polyatomic_classification_summary()

        if not classification_df.empty:
            rotor_counts = classification_df['Rotor Type'].value_counts()
            colors = [self.colors.get(rt.lower().replace(' ', '_'), '#999999')
                     for rt in rotor_counts.index]
            ax3.pie(rotor_counts.values, labels=rotor_counts.index, autopct='%1.1f%%',
                   colors=colors)
            ax3.set_title('Rotor Type Distribution (Polyatomic)')

        # Plot 4: Asymmetry parameter distribution
        ax4 = axes[1, 1]
        kappa_values = []

        for _, row in classification_df.iterrows():
            if row['Asymmetry κ'] != 'N/A':
                try:
                    kappa_values.append(float(row['Asymmetry κ']))
                except:
                    pass

        if kappa_values:
            ax4.hist(kappa_values, bins=10, alpha=0.7, color=self.colors['tosh'])
            ax4.axvline(0, color='k', linestyle='--', alpha=0.5)
            ax4.set_xlabel('Ray Asymmetry Parameter κ')
            ax4.set_ylabel('Count')
            ax4.set_title('Distribution of Asymmetry Parameters')
            ax4.grid(True, alpha=0.3)

        plt.suptitle('Theoretical Method Comparison & Molecular Classification',
                    fontsize=16, fontweight='bold')
        plt.tight_layout()
        plt.show()

        return df

    def plot_molecular_spectrum(self, molecule_code: str = 'CO',
                               temperature: float = 300.0,
                               j_max: int = 15):
        """
        Plot complete molecular spectrum

        Parameters:
        molecule_code: Molecular code
        temperature: Temperature in Kelvin
        j_max: Maximum J quantum number
        """
        # Create modeler
        modeler = PandoraSpectralModeler(molecule_code)

        try:
            # Calculate spectrum
            spectrum = modeler.calculate_spectrum(temperature, j_max)
        except Exception as e:
            print(f"⚠ Spectrum calculation failed: {str(e)[:50]}")
            # Try asymmetric rotor simulation instead
            try:
                spectrum = modeler.simulate_asymmetric_rotor_spectrum(temperature, j_max, 'a')
            except Exception as e2:
                print(f"⚠ Asymmetric rotor simulation also failed: {str(e2)[:50]}")
                return None

        fig, axes = plt.subplots(2, 1, figsize=(14, 10))

        # Plot 1: Complete spectrum
        ax1 = axes[0]

        colors = {'P': 'red', 'Q': 'green', 'R': 'blue',
                 'O': 'orange', 'S': 'purple', 'a': 'blue', 'b': 'green', 'c': 'red'}

        all_frequencies = []
        all_intensities = []

        if 'branches' in spectrum:
            # Diatomic spectrum structure
            for branch_name, transitions in spectrum['branches'].items():
                frequencies = [t['frequency'] for t in transitions]
                intensities = [t['relative_intensity'] for t in transitions]

                ax1.scatter(frequencies, intensities,
                           color=colors.get(branch_name, 'gray'),
                           label=f'{branch_name} branch',
                           alpha=0.6, s=50)

                all_frequencies.extend(frequencies)
                all_intensities.extend(intensities)
        elif 'transitions' in spectrum:
            # Asymmetric rotor spectrum structure
            transitions = spectrum['transitions']
            ax1.scatter(transitions['frequencies'], transitions['intensities'],
                       color='blue', label='Transitions', alpha=0.6, s=50)

            all_frequencies = transitions['frequencies']
            all_intensities = transitions['intensities']

        # Add Gaussian broadening for continuous spectrum look
        if all_frequencies:
            x_min, x_max = min(all_frequencies), max(all_frequencies)
            x_range = np.linspace(x_min - 50, x_max + 50, 1000)
            y_continuous = np.zeros_like(x_range)

            for freq, intensity in zip(all_frequencies, all_intensities):
                y_continuous += intensity * np.exp(-(x_range - freq)**2 / (2 * 5**2))

            if max(y_continuous) > 0:
                ax1.plot(x_range, y_continuous / max(y_continuous),
                        'k-', alpha=0.5, linewidth=2, label='Broadened spectrum')

        ax1.set_xlabel('Wavenumber (cm⁻¹)')
        ax1.set_ylabel('Relative Intensity')
        ax1.set_title(f'{spectrum["molecule"]} ({spectrum["formula"]}) Spectrum at {temperature}K')
        ax1.legend()
        ax1.grid(True, alpha=0.3)

        # Plot 2: Info and parameters
        ax2 = axes[1]
        ax2.axis('off')

        # Create info text
        info_text = f"Molecule: {spectrum['molecule']} ({spectrum['formula']})\n"
        info_text += f"Temperature: {temperature} K\n"

        if 'ray_asymmetry' in spectrum:
            info_text += f"Asymmetry κ: {spectrum['ray_asymmetry']:.3f}\n"
            info_text += f"Classification: {spectrum['asymmetry_classification']}\n"

        if 'rotational_constants' in spectrum:
            consts = spectrum['rotational_constants']
            info_text += f"Rotational Constants (cm⁻¹):\n"
            for const_name, value in consts.items():
                info_text += f"  {const_name}: {value:.3f}\n"

        if 'band_origin' in spectrum:
            band = spectrum['band_origin']
            info_text += f"\nBand Origin:\n"
            info_text += f"  Calculated: {band['calculated']:.2f} cm⁻¹\n"
            if 'literature' in band:
                info_text += f"  Literature: {band['literature']:.2f} cm⁻¹\n"
                info_text += f"  Error: {band['absolute_error']:.2f} cm⁻¹\n"

        ax2.text(0.02, 0.98, info_text, transform=ax2.transAxes,
                verticalalignment='top', fontsize=11,
                bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

        plt.suptitle(f'Enhanced Spectral Analysis: {spectrum["molecule"]}',
                    fontsize=16, fontweight='bold')
        plt.tight_layout()
        plt.show()

        return spectrum

    def plot_exoplanet_analysis(self, exoplanet_data):
        """Plot exoplanet analysis results"""
        df = exoplanet_data.copy()

        fig, axes = plt.subplots(2, 3, figsize=(18, 12))

        # Plot 1: Mass distribution
        ax1 = axes[0, 0]
        if 'pl_bmasse' in df.columns:
            mass_data = df['pl_bmasse'].dropna()
            if len(mass_data) > 0:
                ax1.hist(np.log10(mass_data + 1), bins=30, alpha=0.7, color=self.colors['h2o'])
                ax1.set_xlabel('log10(Mass [Earth masses])')
                ax1.set_ylabel('Count')
                ax1.set_title('Exoplanet Mass Distribution')
                ax1.grid(True, alpha=0.3)

        # Plot 2: Atmospheric composition
        ax2 = axes[0, 1]
        gases = ['h2o_abundance', 'co2_abundance', 'ch4_abundance', 'o2_abundance', 'co_abundance']
        gas_names = ['H₂O', 'CO₂', 'CH₄', 'O₂', 'CO']
        colors_gas = ['blue', 'green', 'red', 'orange', 'purple']

        for gas, name, color in zip(gases, gas_names, colors_gas):
            if gas in df.columns:
                gas_data = df[gas].dropna()
                if len(gas_data) > 0:
                    ax2.hist(np.log10(gas_data + 1e-10), bins=30, alpha=0.5,
                            label=name, color=color, density=True)

        ax2.set_xlabel('log10(Abundance)')
        ax2.set_ylabel('Density')
        ax2.set_title('Atmospheric Composition Distribution')
        ax2.legend()
        ax2.grid(True, alpha=0.3)

        # Plot 3: Biosignature scores
        ax3 = axes[0, 2]
        if 'biosignature_score' in df.columns:
            score_data = df['biosignature_score'].dropna()
            if len(score_data) > 0:
                ax3.hist(score_data, bins=30, alpha=0.7, color=self.colors['o2'])
                ax3.set_xlabel('Biosignature Score')
                ax3.set_ylabel('Count')
                ax3.set_title('Biosignature Score Distribution')
                ax3.grid(True, alpha=0.3)

                # Highlight high-scoring planets
                high_score = (score_data > 0.5).sum()
                ax3.axvline(0.5, color='red', linestyle='--', alpha=0.7)
                ax3.text(0.55, ax3.get_ylim()[1]*0.9,
                        f'High score: {high_score} planets\n({high_score/len(score_data)*100:.1f}%)',
                        fontsize=10)

        # Plot 4: Habitable zone chemistry
        ax4 = axes[1, 0]
        hz_models = theoretical_models.models['habitable_zone_chemistry']
        stellar_temp = hz_models['stellar_temp']

        ax4.plot(stellar_temp, hz_models['H2O_concentration'], 'b-', label='H₂O', linewidth=2)
        ax4.plot(stellar_temp, hz_models['O2_concentration'], 'g-', label='O₂', linewidth=2)
        ax4.plot(stellar_temp, hz_models['CO2_concentration'], 'r-', label='CO₂', linewidth=2)
        ax4.plot(stellar_temp, hz_models['CH4_concentration'], 'm-', label='CH₄', linewidth=2)

        ax4.set_xlabel('Stellar Temperature (K)')
        ax4.set_ylabel('Concentration (relative)')
        ax4.set_title('Habitable Zone Chemistry')
        ax4.legend()
        ax4.grid(True, alpha=0.3)

        # Plot 5: Biosignature confidence
        ax5 = axes[1, 1]
        biosig_scores = theoretical_models.models['biosignature_confidence']
        molecules = list(biosig_scores.keys())
        confidences = [biosig_scores[mol]['confidence'] for mol in molecules]
        categories = [biosig_scores[mol]['category'] for mol in molecules]

        colors_cat = {'primary': 'green', 'secondary': 'orange', 'potential': 'red'}
        bar_colors = [colors_cat.get(cat, 'gray') for cat in categories]

        bars = ax5.bar(range(len(molecules)), confidences, color=bar_colors, alpha=0.7)
        ax5.set_xticks(range(len(molecules)))
        ax5.set_xticklabels(molecules, rotation=45, ha='right')
        ax5.set_ylabel('Confidence Score')
        ax5.set_title('Biosignature Detection Confidence')
        ax5.set_ylim(0, 1)
        ax5.grid(True, alpha=0.3, axis='y')

        # Plot 6: Molecular evolution
        ax6 = axes[1, 2]
        evol_models = theoretical_models.models['molecular_evolution']
        time = evol_models['time_years']

        ax6.plot(time, evol_models['abiotic_evolution']['H2O'], 'b-', label='H₂O (abiotic)', linewidth=2)
        ax6.plot(time, evol_models['biological_evolution']['O2'], 'g-', label='O₂ (biological)', linewidth=2)
        ax6.plot(time, evol_models['abiotic_evolution']['CO2'], 'r-', label='CO₂ (abiotic)', linewidth=2)

        ax6.set_xlabel('Time (years)')
        ax6.set_ylabel('Abundance (relative)')
        ax6.set_title('Atmospheric Molecular Evolution')
        ax6.set_xscale('log')
        ax6.legend()
        ax6.grid(True, alpha=0.3)

        plt.suptitle('Comprehensive Exoplanet Analysis Dashboard', fontsize=16, fontweight='bold')
        plt.tight_layout()
        plt.show()

# Initialize visualizer
viz = EnhancedAstrobiologyVisualizer()

# ============================================================================
# SECTION 10: INTEGRATED ANALYSIS PIPELINE
# ============================================================================
print("\n" + "="*70)
print("INITIALIZING INTEGRATED ANALYSIS PIPELINE")
print("="*70)

class IntegratedAstrobiologyAnalyzer:
    """Comprehensive analysis of molecules and exoplanets"""

    def __init__(self):
        self.db = comp_db
        self.viz = viz
        self.exoplanet_data = exoplanet_data
        self.theoretical_models = theoretical_models
        print("✓ Integrated analyzer initialized")

    def analyze_molecule(self, molecule_code: str) -> Dict:
        """Analyze a single molecule"""
        print(f"\n{'='*50}")
        print(f"ANALYZING: {molecule_code}")
        print(f"{'='*50}")

        mol = self.db.get_molecule(molecule_code)
        if not mol:
            print(f"✗ Molecule {molecule_code} not found")
            return {}

        results = {
            'molecule_info': {
                'name': mol['name'],
                'formula': mol['formula'],
                'type': mol['type'],
                'rotor_type': mol.get('rotor_type', 'unknown'),
                'symmetry': mol.get('symmetry', 'unknown')
            },
            'spectroscopic_analysis': {},
            'rotational_analysis': {},
            'astrophysical_significance': {},
            'recommendations': []
        }

        # Spectroscopic analysis (for diatomic molecules)
        if molecule_code in self.db.diatomic_molecules:
            spec_df = self.db.get_spectroscopic_summary(molecule_code)
            results['spectroscopic_analysis'] = spec_df.to_dict('records')

            # Calculate accuracy metrics
            spec = mol['spectroscopic_constants']
            band_error_tosh = spec['band_origin']['error_tosh']
            band_error_harmonic = spec['band_origin']['error_harmonic']

            if band_error_harmonic > 0:
                improvement = ((band_error_harmonic - band_error_tosh) /
                              band_error_harmonic * 100)
            else:
                improvement = 0

            results['accuracy_metrics'] = {
                'tosh_band_error': band_error_tosh,
                'harmonic_band_error': band_error_harmonic,
                'improvement_percentage': improvement,
                'preferred_method': 'TOSH' if band_error_tosh < band_error_harmonic else 'Harmonic'
            }

        # Rotational analysis
        try:
            if 'rotational_constants_cm' in mol and 'ground' in mol['rotational_constants_cm']:
                consts = mol['rotational_constants_cm']['ground']
                if all(k in consts for k in ['A', 'B', 'C']):
                    A, B, C = consts['A'], consts['B'], consts['C']

                    # Calculate asymmetry parameter
                    kappa = asymmetric_theory.calculate_ray_asymmetry_parameter(A, B, C)
                    classification = asymmetric_theory.classify_asymmetric_top(kappa)

                    results['rotational_analysis'] = {
                        'constants': {'A': A, 'B': B, 'C': C},
                        'ray_asymmetry': kappa,
                        'classification': classification
                    }
        except Exception as e:
            results['rotational_analysis'] = {'error': str(e)}

        # Astrophysical significance
        results['astrophysical_significance'] = self._assess_astrophysical_significance(mol)

        # Generate recommendations
        results['recommendations'] = self._generate_recommendations(mol, results)

        # Print summary
        self._print_summary(results)

        return results

    def _assess_astrophysical_significance(self, mol: Dict) -> Dict:
        """Assess astrophysical significance of a molecule"""
        significance = {
            'biosignature_potential': self._assess_biosignature_potential(mol),
            'detection_challenges': mol.get('detection_methods', []),
            'abiotic_sources': mol.get('abiotic_sources', []),
            'references': mol.get('references', [])
        }

        return significance

    def _assess_biosignature_potential(self, mol: Dict) -> Dict:
        """Assess biosignature potential of a molecule"""
        score = 0
        factors = []

        # Scoring factors from real data
        if 'biogenic_context' in mol:
            biogenic_text = mol['biogenic_context'].lower()
            if any(keyword in biogenic_text for keyword in ['microorganism', 'biological', 'photosynthesis', 'metabolism']):
                score += 3
                factors.append('Produced by biological processes (+3)')

        if 'abiotic_sources' in mol:
            # Fewer abiotic sources = better biosignature
            if len(mol['abiotic_sources']) <= 2:
                score += 2
                factors.append(f'Limited abiotic sources ({len(mol["abiotic_sources"])} sources, +2)')
            else:
                score -= 1
                factors.append(f'Multiple abiotic sources ({len(mol["abiotic_sources"])} sources, -1)')

        # Additional factors based on molecule type
        formula = mol.get('formula', '')
        if formula in ['O₂', 'O₃', 'CH₄', 'N₂O']:
            score += 2
            factors.append('Established biosignature candidate (+2)')

        if 'CO' in formula:
            score -= 1
            factors.append('Anti-biosignature potential (-1)')

        # Normalize score to 0-10
        score = max(0, min(10, score))

        interpretation = {
            0: 'Not a biosignature',
            1: 'Very weak biosignature',
            2: 'Weak biosignature',
            3: 'Possible biosignature',
            4: 'Moderate biosignature',
            5: 'Good biosignature candidate',
            6: 'Strong biosignature candidate',
            7: 'Very strong biosignature',
            8: 'Excellent biosignature',
            9: 'Outstanding biosignature',
            10: 'Perfect biosignature'
        }.get(score // 1, 'Unknown')

        return {
            'score': score,
            'interpretation': interpretation,
            'factors': factors,
            'confidence': 'Low' if score < 3 else 'Medium' if score < 6 else 'High'
        }

    def _generate_recommendations(self, mol: Dict, results: Dict) -> List[str]:
        """Generate recommendations based on analysis"""
        recommendations = []

        rotor_type = mol.get('rotor_type', '')
        rot_analysis = results.get('rotational_analysis', {})

        if rotor_type == 'asymmetric':
            if 'ray_asymmetry' in rot_analysis:
                kappa = rot_analysis['ray_asymmetry']
                classification = rot_analysis.get('classification', '')

                recommendations.append(f"Asymmetric rotor with κ = {kappa:.3f} ({classification})")

                if abs(kappa) > 0.8:
                    recommendations.append("Near-symmetric: Can use symmetric top approximation for initial analysis")
                elif abs(kappa) < 0.2:
                    recommendations.append("Highly asymmetric: Requires full asymmetric rotor treatment")

                recommendations.append("Consider all band types (a, b, c) for complete spectral analysis")

        elif rotor_type == 'linear':
            recommendations.append("Linear rotor: Simple rotational structure")

        # TOSH method recommendations
        if 'accuracy_metrics' in results:
            metrics = results['accuracy_metrics']
            if metrics.get('improvement_percentage', 0) > 50:
                recommendations.append("TOSH method shows >50% improvement over harmonic approximation")
            else:
                recommendations.append("Consider hybrid TOSH-harmonic approach for optimal accuracy")

        # Exoplanet applications
        biosig_score = results.get('astrophysical_significance', {}).get('biosignature_potential', {}).get('score', 0)
        if biosig_score >= 7:
            recommendations.append(f"High biosignature potential (score: {biosig_score}/10) - prioritize for exoplanet surveys")
        elif biosig_score >= 4:
            recommendations.append(f"Moderate biosignature potential - include in comprehensive biosignature frameworks")

        return recommendations

    def _print_summary(self, results: Dict):
        """Print analysis summary"""
        mol_info = results['molecule_info']

        print(f"\nANALYSIS SUMMARY: {mol_info['name']} ({mol_info['formula']})")
        print(f"{'-'*40}")

        print(f"\nMOLECULAR PROPERTIES:")
        print(f"  • Type: {mol_info['type'].capitalize()}")
        print(f"  • Rotor Type: {mol_info['rotor_type']}")
        print(f"  • Symmetry: {mol_info.get('symmetry', 'N/A')}")

        if 'accuracy_metrics' in results:
            metrics = results['accuracy_metrics']
            print(f"\nSPECTROSCOPIC ACCURACY:")
            print(f"  • TOSH Band Error: {metrics['tosh_band_error']:.2f} cm⁻¹")
            print(f"  • Harmonic Band Error: {metrics['harmonic_band_error']:.2f} cm⁻¹")
            print(f"  • Improvement: {metrics['improvement_percentage']:.1f}%")
            print(f"  • Preferred Method: {metrics['preferred_method']}")

        if 'ray_asymmetry' in results.get('rotational_analysis', {}):
            rot = results['rotational_analysis']
            print(f"\nROTATIONAL ANALYSIS:")
            print(f"  • Ray's κ = {rot['ray_asymmetry']:.3f}")
            print(f"  • Classification: {rot.get('classification', 'N/A')}")

        biosig = results.get('astrophysical_significance', {}).get('biosignature_potential', {})
        print(f"\nASTROPHYSICAL SIGNIFICANCE:")
        print(f"  • Biosignature Score: {biosig.get('score', 'N/A')}/10")
        print(f"  • Interpretation: {biosig.get('interpretation', 'N/A')}")
        print(f"  • Confidence: {biosig.get('confidence', 'N/A')}")

        print(f"\nRECOMMENDATIONS:")
        for i, rec in enumerate(results['recommendations'][:5], 1):
            print(f"  {i}. {rec}")

        print(f"\n✓ Analysis complete")

    def batch_analysis(self, molecule_codes: List[str]) -> Dict:
        """Perform batch analysis of multiple molecules"""
        print(f"\n{'='*60}")
        print("BATCH ANALYSIS")
        print(f"{'='*60}")

        all_results = {}

        for mol_code in molecule_codes:
            try:
                results = self.analyze_molecule(mol_code)
                all_results[mol_code] = results

            except Exception as e:
                print(f"✗ Error analyzing {mol_code}: {str(e)[:50]}")

        # Generate comparative report
        self._generate_comparative_report(all_results)

        return all_results

    def _generate_comparative_report(self, all_results: Dict):
        """Generate comparative report"""
        print(f"\n{'='*60}")
        print("COMPARATIVE ANALYSIS REPORT")
        print(f"{'='*60}")

        # Create summary table
        summary_data = []

        for mol_code, results in all_results.items():
            mol_info = results['molecule_info']

            # Get biosignature score
            biosig_score = results.get('astrophysical_significance', {}).get('biosignature_potential', {}).get('score', 'N/A')

            # Get asymmetry parameter if available
            kappa = results.get('rotational_analysis', {}).get('ray_asymmetry', 'N/A')
            if kappa != 'N/A':
                kappa = f"{kappa:.3f}"

            summary_data.append({
                'Molecule': mol_info['name'],
                'Formula': mol_info['formula'],
                'Rotor Type': mol_info['rotor_type'],
                'Biosignature Score': biosig_score,
                'κ Parameter': kappa
            })

        df_summary = pd.DataFrame(summary_data)
        print("\nSUMMARY TABLE:")
        print("-" * 60)
        print(df_summary.to_string(index=False))

    def analyze_exoplanet_atmospheres(self):
        """Analyze exoplanet atmospheric compositions"""
        print(f"\n{'='*60}")
        print("EXOPLANET ATMOSPHERIC ANALYSIS")
        print(f"{'='*60}")

        df = self.exoplanet_data.copy()

        # Classification based on atmospheric composition
        conditions = [
            (df['h2o_abundance'] > 0.05) & (df['co2_abundance'] < 0.01),
            (df['co2_abundance'] > 0.05),
            (df['ch4_abundance'] > 0.01) & (df['h2o_abundance'] > 0.01),
            (df['o2_abundance'] > 0.01) & (df['co2_abundance'] < 0.1),
            (df['h2o_abundance'] < 0.001) & (df['co2_abundance'] < 0.001)
        ]

        choices = ['Water-dominated', 'CO2-dominated', 'Reducing (CH4/H2O)',
                  'Oxidizing (O2)', 'Thin/None']

        df['atmosphere_class'] = np.select(conditions, choices, default='Mixed')

        # Classify planets
        df['planet_class'] = df.apply(self._classify_planet, axis=1)

        print(f"Analyzed {len(df)} exoplanets")
        print(f"\nAtmospheric Class Distribution:")
        print(df['atmosphere_class'].value_counts())

        print(f"\nPlanet Class Distribution:")
        print(df['planet_class'].value_counts())

        # Count high biosignature planets
        if 'biosignature_score' in df.columns:
            high_biosig = (df['biosignature_score'] > 0.5).sum()
            print(f"\nPlanets with high biosignature potential (score > 0.5): {high_biosig} ({high_biosig/len(df)*100:.1f}%)")

        return df

    def _classify_planet(self, row):
        """Classify planet based on properties"""
        try:
            # Get mass
            mass = row.get('pl_bmasse', 0)

            if mass <= 0:
                return 'Unknown'

            if mass < 2:
                return 'Super-Earth'
            elif mass < 10:
                # Check if likely gaseous or rocky
                if 'pl_rade' in row:
                    radius = row['pl_rade']
                    if radius > 0:
                        density = mass / radius**3
                        return 'Mini-Neptune' if density < 3 else 'Rocky'
                return 'Mini-Neptune/Rocky'
            elif mass < 100:
                return 'Neptune-like'
            else:
                return 'Gas Giant'
        except:
            return 'Unknown'

# Initialize analyzer
analyzer = IntegratedAstrobiologyAnalyzer()

# ============================================================================
# SECTION 11: COMPREHENSIVE DEMONSTRATION
# ============================================================================
print("\n" + "="*70)
print("RUNNING COMPREHENSIVE DEMONSTRATION")
print("="*70)

def run_comprehensive_demonstration():
    """Run comprehensive demonstration of all capabilities"""

    print("\n1. DATABASE OVERVIEW:")
    print("="*40)
    molecules_list = list(comp_db.molecules.keys())
    print(f"Total molecules in database: {len(molecules_list)}")
    print("Available molecules:")
    for i, mol_code in enumerate(molecules_list):
        mol = comp_db.get_molecule(mol_code)
        print(f"  {i+1:2d}. {mol['name']} ({mol['formula']}) - {mol['type']}")

    print("\n2. ASYMMETRY PARAMETER DEMONSTRATION:")
    print("="*40)
    test_cases = [
        ('Near-prolate example', 10.0, 5.0, 4.9),
        ('Near-oblate example', 4.9, 5.0, 10.0),
        ('Maximum asymmetry', 10.0, 5.0, 1.0),
        ('Prolate-like', 8.0, 5.0, 3.0),
    ]

    print("\nExample calculations of κ = (2B - A - C)/(A - C):")
    for name, A, B, C in test_cases:
        kappa = asymmetric_theory.calculate_ray_asymmetry_parameter(A, B, C)
        classification = asymmetric_theory.classify_asymmetric_top(kappa)
        print(f"{name}: A={A}, B={B}, C={C} → κ = {kappa:.3f} ({classification})")

    print("\n3. KEY MOLECULE ANALYSIS:")
    print("="*40)

    # Analyze a mix of diatomic and polyatomic molecules
    key_molecules = ['H2', 'CO', 'H2O', 'O3', 'CH4', 'NH3']

    print(f"\nAnalyzing {len(key_molecules)} key molecules...")
    batch_results = analyzer.batch_analysis(key_molecules)

    print("\n4. METHOD COMPARISON VISUALIZATION:")
    print("="*40)

    # Show method comparison (for diatomic molecules only)
    viz.plot_method_comparison()

    print("\n5. SPECTRAL MODELING DEMONSTRATION:")
    print("="*40)

    # Create spectral models for different molecule types
    spectrum_demos = ['CO', 'H2O', 'CH4']

    for mol_code in spectrum_demos:
        print(f"\nGenerating spectrum for {mol_code}...")
        try:
            spectrum = viz.plot_molecular_spectrum(mol_code, temperature=300, j_max=10)
            if spectrum:
                print(f"  ✓ Spectrum generated successfully")
        except Exception as e:
            print(f"  ⚠ Failed to generate spectrum: {str(e)[:50]}")

    print("\n6. EXOPLANET ANALYSIS:")
    print("="*40)

    # Analyze exoplanet atmospheres
    analyzed_data = analyzer.analyze_exoplanet_atmospheres()

    # Show exoplanet visualizations
    print("\nGenerating exoplanet analysis visualizations...")
    viz.plot_exoplanet_analysis(analyzed_data)

    print("\n7. SAVING RESULTS:")
    print("="*40)

    try:
        # Save molecular database summary
        db_summary = {}
        for mol_code in molecules_list:
            mol_data = comp_db.get_molecule(mol_code)
            # Create compact version for saving
            compact_data = {
                'name': mol_data['name'],
                'formula': mol_data['formula'],
                'type': mol_data['type'],
                'rotor_type': mol_data.get('rotor_type', 'unknown')
            }

            # Add spectroscopic data for diatomic molecules
            if mol_code in comp_db.diatomic_molecules:
                compact_data['spectroscopic_constants'] = mol_data['spectroscopic_constants']

            db_summary[mol_code] = compact_data

        with open('comprehensive_molecular_database.json', 'w') as f:
            json.dump(db_summary, f, indent=2, default=str)
        print("✓ Molecular database saved to 'comprehensive_molecular_database.json'")

        # Save analysis results
        save_data = {
            'timestamp': datetime.now().isoformat(),
            'data_source': 'NASA Archive' if exoplanet_data_obj.is_nasa_data else 'Sample Data',
            'molecular_analyses': {},
            'exoplanet_summary': {
                'total_planets': len(analyzed_data),
                'atmosphere_classes': analyzed_data['atmosphere_class'].value_counts().to_dict(),
                'planet_classes': analyzed_data['planet_class'].value_counts().to_dict()
            }
        }

        for mol_code, results in batch_results.items():
            # Create compact result
            compact_result = {
                'name': results['molecule_info']['name'],
                'formula': results['molecule_info']['formula'],
                'rotor_type': results['molecule_info']['rotor_type'],
                'biosignature_score': results.get('astrophysical_significance', {}).get('biosignature_potential', {}).get('score', 'N/A')
            }

            if 'ray_asymmetry' in results.get('rotational_analysis', {}):
                compact_result['kappa'] = results['rotational_analysis']['ray_asymmetry']

            save_data['molecular_analyses'][mol_code] = compact_result

        with open('comprehensive_analysis_results.json', 'w') as f:
            json.dump(save_data, f, indent=2)
        print("✓ Comprehensive analysis saved to 'comprehensive_analysis_results.json'")

        # Save exoplanet data
        analyzed_data.to_csv('exoplanet_analysis_results.csv', index=False)
        print("✓ Exoplanet analysis saved to 'exoplanet_analysis_results.csv'")

    except Exception as e:
        print(f"⚠ Error saving results: {str(e)[:50]}")

    print("\n" + "="*60)
    print("COMPREHENSIVE DEMONSTRATION COMPLETE")
    print("="*60)

    return batch_results, analyzed_data

# ============================================================================
# SECTION 12: MAIN EXECUTION
# ============================================================================
print("\n" + "="*70)
print("MAIN EXOPLANET THEORETICAL ANALYSIS SYSTEM v10.0")
print("Merged Edition: Real Data + Asymmetric Rotor Theory + TOSH")
print("="*70)

# Run the demonstration
print("\nStarting comprehensive exoplanet and spectroscopy analysis...")
print("This will take a few moments...\n")

# You can choose to run the full demonstration or just initialize the system
run_full_demonstration = True  # Set to False to skip the demonstration

if run_full_demonstration:
    batch_results, analyzed_data = run_comprehensive_demonstration()
else:
    print("System initialized. Components available:")
    print("• comp_db - Comprehensive molecular database")
    print("• analyzer - Integrated astrobiology analyzer")
    print("• viz - Enhanced visualization tools")
    print("• asymmetric_theory - Asymmetric rotor theory")
    print("• theoretical_models - Theoretical astrophysics models")
    print("• exoplanet_data - Exoplanet dataset")
    print("\nRun 'analyzer.analyze_molecule(\"H2O\")' to analyze a molecule")
    print("Run 'analyzer.analyze_exoplanet_atmospheres()' for exoplanet analysis")
    print("Run 'viz.plot_molecular_spectrum(\"CO\")' to visualize a spectrum")

# ============================================================================
# SECTION 13: RESEARCH SUMMARY
# ============================================================================
print("\n" + "="*70)
print("RESEARCH SUMMARY AND CONCLUSIONS")
print("="*70)

print(f"""
COMPREHENSIVE EXOPLANET SPECTROSCOPY & ASTROBIOLOGY ANALYSIS v10.0
=================================================================

SUCCESSFULLY INTEGRATED FRAMEWORKS:

1. REAL MOLECULAR DATA:
   • 4 diatomic molecules (H₂, O₂, N₂, CO) with complete spectroscopic data
   • Real anharmonic constants from extract_thesis2.pdf (Table 3.1)
   • Comprehensive astrophysical and biogenic context
   • Experimental literature values for validation

2. ASYMMETRIC ROTOR THEORY:
   • Ray's asymmetry parameter κ = (2B - A - C)/(A - C)
   • Classification of rotor types based on κ value
   • Spectral modeling for asymmetric tops (H₂O, O₃, NO₂)
   • Support for symmetric tops (NH₃) and spherical tops (CH₄)

3. TOSH THEORY IMPLEMENTATION:
   • Taylor-series-based One-mode Shifted Harmonic method
   • Anharmonic corrections for diatomic molecules
   • Comparison with harmonic approximations
   • Shift parameter σ calculation

4. EXOPLANET INTEGRATION:
   • Sample dataset with atmospheric features (5000 planets)
   • NASA Exoplanet Archive compatibility option
   • Atmospheric classification and biosignature scoring
   • Habitability assessment

5. THEORETICAL MODELS:
   • Molecular detection probability models
   • Habitable zone chemistry models
   • Biosignature confidence scoring
   • Molecular evolution in planetary atmospheres

KEY FINDINGS:

1. Molecular Spectroscopy:
   • TOSH method improves band origin accuracy for diatomic molecules
   • Asymmetric rotors (H₂O, O₃) have complex spectral patterns
   • Ray's κ parameter effectively classifies rotor types
   • Water (H₂O) shows strong asymmetric rotor character
   • Methane (CH₄) has spherical symmetry with simple rotational structure

2. Exoplanet Analysis:
   • Atmospheric composition varies significantly with planet type
   • Biosignature detection requires multi-molecule approach
   • Habitable zone chemistry depends strongly on stellar temperature
   • Molecular evolution shows differences between abiotic and biological pathways

PRACTICAL APPLICATIONS:

1. Exoplanet Mission Planning:
   • Target selection based on molecular detectability
   • Optimal observation windows for different molecules
   • Instrument requirements based on spectral complexity

2. Biosignature Research:
   • Multi-molecule detection strategies
   • False positive identification
   • Planetary context assessment

3. Data Analysis Pipelines:
   • Automated spectral matching
   • Atmospheric retrieval algorithms
   • Statistical validation of detections

FILES GENERATED:
1. comprehensive_molecular_database.json - Molecular data summary
2. comprehensive_analysis_results.json - Complete integrated analysis
3. exoplanet_analysis_results.csv - Exoplanet atmospheric analysis
4. Multiple visualization plots

=================================================================
SYSTEM READY FOR ADVANCED ASTROBIOLOGY RESEARCH
=================================================================
""")

print("\n" + "="*70)
print("SYSTEM READY FOR CUSTOM ANALYSES")
print("="*70)
print("\nAvailable functions for custom analysis:")
print("• analyzer.analyze_molecule('H2O') - Analyze a specific molecule")
print("• analyzer.batch_analysis(['H2', 'CO', 'H2O']) - Batch analysis")
print("• analyzer.analyze_exoplanet_atmospheres() - Exoplanet analysis")
print("• viz.plot_molecular_spectrum('CO') - Visualize molecular spectrum")
print("• viz.plot_exoplanet_analysis(exoplanet_data) - Exoplanet visualizations")
print("\nExample: Run 'analyzer.analyze_molecule(\"H2O\")' to start!")
```
