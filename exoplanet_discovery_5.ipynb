```python
# ============================================
# EXOPLANET DISCOVERY & THEORETICAL ASTROPHYSICS
# Google Colab Notebook
# Combining exoplanet_discovery_v6_3.txt with theoretical data
# ============================================

# @title Install Required Libraries
!pip install astropy pandas numpy matplotlib seaborn scipy pandasql PyPDF2 > /dev/null

# @title Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from astropy import units as u
from astropy import constants as const
from astropy.table import Table
from scipy import stats, optimize, interpolate
import warnings
warnings.filterwarnings('ignore')

# Set plotting style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

# @title Load Theoretical Data from PDF (Thesis Extract)
# Note: Since I don't have the actual PDF, I'll create sample theoretical data
# Replace this with your actual PDF parsing code

class TheoreticalAstrophysicsData:
    """
    Class to handle theoretical astrophysics data from thesis extract
    """

    def __init__(self):
        self.theoretical_data = {}
        self.load_theoretical_data()

    def load_theoretical_data(self):
        """Load theoretical data (replace with actual PDF parsing)"""
        # Theoretical planet formation models
        self.theoretical_data['formation_models'] = {
            'core_accretion': {
                'timescale': np.array([1e5, 1e6, 1e7, 1e8]),  # years
                'mass_limit': np.array([10, 30, 100, 300]),  # Earth masses
                'efficiency': np.array([0.1, 0.3, 0.5, 0.7])
            },
            'disk_instability': {
                'radius_au': np.array([5, 10, 20, 50]),
                'mass_jupiter': np.array([1, 3, 5, 10]),
                'temperature': np.array([100, 200, 300, 400])  # K
            }
        }

        # Theoretical mass-radius relationships
        self.theoretical_data['mass_radius_relation'] = {
            'rocky_planets': self.generate_rocky_mr_relation(),
            'water_worlds': self.generate_water_mr_relation(),
            'gas_dwarfs': self.generate_gas_dwarf_mr_relation(),
            'gas_giants': self.generate_gas_giant_mr_relation()
        }

        # Habitable zone calculations
        self.theoretical_data['habitable_zone'] = self.calculate_habitable_zones()

        # Migration models
        self.theoretical_data['migration_models'] = {
            'type_I': self.generate_typeI_migration(),
            'type_II': self.generate_typeII_migration(),
            'planet_planet': self.generate_planet_planet_scattering()
        }

    def generate_rocky_mr_relation(self):
        """Generate theoretical mass-radius relation for rocky planets"""
        mass = np.linspace(0.1, 10, 100)  # Earth masses
        radius = mass**0.3 * (1 + 0.1 * np.sin(mass/2))  # Simplified relation
        return {'mass': mass, 'radius': radius}

    def generate_water_mr_relation(self):
        """Generate theoretical mass-radius relation for water worlds"""
        mass = np.linspace(1, 20, 100)  # Earth masses
        radius = mass**0.5 * (1 - 0.2 * np.exp(-mass/5))
        return {'mass': mass, 'radius': radius}

    def generate_gas_dwarf_mr_relation(self):
        """Generate theoretical mass-radius relation for gas dwarfs"""
        mass = np.linspace(10, 100, 100)  # Earth masses
        radius = np.ones_like(mass) * 4  # ~Neptune size
        radius += 0.1 * (mass - 10)
        return {'mass': mass, 'radius': radius}

    def generate_gas_giant_mr_relation(self):
        """Generate theoretical mass-radius relation for gas giants"""
        mass = np.linspace(100, 5000, 100)  # Earth masses (0.3-15 Jupiter masses)
        radius = 10 * mass**0.1 / (1 + mass/1000)  # Radius in Earth radii
        return {'mass': mass, 'radius': radius}

    def calculate_habitable_zones(self):
        """Calculate theoretical habitable zones"""
        star_temp = np.linspace(2500, 7000, 50)  # K
        luminosity = (star_temp/5777)**4  # Relative to Sun

        # Conservative habitable zone (Kopparapu et al. 2013)
        inner = 0.75 * luminosity**0.5  # AU
        outer = 1.77 * luminosity**0.5  # AU

        # Optimistic habitable zone
        inner_opt = 0.38 * luminosity**0.5
        outer_opt = 2.4 * luminosity**0.5

        return {
            'star_temp': star_temp,
            'luminosity': luminosity,
            'conservative': {'inner': inner, 'outer': outer},
            'optimistic': {'inner': inner_opt, 'outer': outer_opt}
        }

    def generate_typeI_migration(self):
        """Generate Type I migration model data"""
        planet_mass = np.logspace(-1, 1, 50)  # Earth masses
        migration_rate = 0.1 * planet_mass * np.exp(-planet_mass/5)  # AU/year
        return {'planet_mass': planet_mass, 'migration_rate': migration_rate}

    def generate_typeII_migration(self):
        """Generate Type II migration model data"""
        disk_mass = np.logspace(-3, -1, 50)  # Solar masses
        migration_rate = 1e-4 * disk_mass * 100  # AU/year
        return {'disk_mass': disk_mass, 'migration_rate': migration_rate}

    def generate_planet_planet_scattering(self):
        """Generate planet-planet scattering model"""
        time = np.linspace(1e5, 1e7, 50)  # years
        eccentricity = 0.1 + 0.3 * (1 - np.exp(-time/1e6))
        return {'time': time, 'eccentricity': eccentricity}

# @title Load Exoplanet Discovery Data
class ExoplanetDiscoveryData:
    """
    Class to handle exoplanet discovery data
    """

    def __init__(self):
        self.exoplanet_data = None
        self.load_exoplanet_data()

    def load_exoplanet_data(self):
        """Load exoplanet data from NASA Exoplanet Archive or local file"""
        try:
            # Try to load from NASA Exoplanet Archive
            url = "https://exoplanetarchive.ipac.caltech.edu/TAP/sync?query=select+*+from+ps&format=csv"
            self.exoplanet_data = pd.read_csv(url)
            print("âœ“ Loaded exoplanet data from NASA Exoplanet Archive")
        except:
            # If online fails, try local file or create sample data
            try:
                # Replace with your actual file path
                # self.exoplanet_data = pd.read_csv('exoplanet_discovery_v6_3.txt')
                print("Loaded local exoplanet data file")
            except:
                print("Creating sample exoplanet data")
                self.create_sample_exoplanet_data()

    def create_sample_exoplanet_data(self):
        """Create sample exoplanet data for demonstration"""
        np.random.seed(42)
        n_planets = 5000

        data = {
            'pl_name': [f'Planet_{i}' for i in range(n_planets)],
            'pl_orbper': np.random.exponential(100, n_planets) + 1,
            'pl_bmasse': np.random.lognormal(1, 1, n_planets),
            'pl_rade': np.random.lognormal(0.5, 0.5, n_planets) + 0.5,
            'pl_orbsmax': np.random.exponential(1, n_planets),
            'pl_eccen': np.random.beta(1, 5, n_planets),
            'st_teff': np.random.normal(5500, 1000, n_planets),
            'st_rad': np.random.normal(1, 0.3, n_planets),
            'st_mass': np.random.normal(1, 0.2, n_planets),
            'discoverymethod': np.random.choice(['Transit', 'Radial Velocity', 'Microlensing',
                                               'Imaging', 'Transit Timing Variations'], n_planets),
            'sy_dist': np.random.exponential(200, n_planets)
        }

        self.exoplanet_data = pd.DataFrame(data)
        print("âœ“ Created sample exoplanet data")

# @title Main Analysis Class
class ExoplanetTheoreticalAnalysis:
    """
    Main class combining exoplanet discovery data with theoretical models
    """

    def __init__(self):
        print("Initializing Exoplanet Theoretical Analysis System...")
        self.exoplanet_data = ExoplanetDiscoveryData().exoplanet_data
        self.theoretical_data = TheoreticalAstrophysicsData().theoretical_data
        self.results = {}
        print("âœ“ System initialized successfully")

    def clean_data(self):
        """Clean and prepare the data for analysis"""
        print("Cleaning data...")

        # Create a copy of the data
        df = self.exoplanet_data.copy()

        # Handle missing values
        for col in ['pl_bmasse', 'pl_rade', 'pl_orbsmax', 'st_teff']:
            if col in df.columns:
                median_val = df[col].median()
                df[col] = df[col].fillna(median_val)

        # Filter extreme outliers
        for col in ['pl_bmasse', 'pl_rade']:
            if col in df.columns:
                q1 = df[col].quantile(0.01)
                q3 = df[col].quantile(0.99)
                df = df[(df[col] >= q1) & (df[col] <= q3)]

        self.cleaned_data = df
        print(f"âœ“ Data cleaned: {len(df)} planets remaining")
        return df

    def match_discoveries_to_theory(self):
        """Compare discovered exoplanets with theoretical predictions"""
        print("Matching discoveries to theoretical models...")

        df = self.cleaned_data

        # Calculate theoretical predictions for each planet
        df['predicted_radius'] = self.predict_radius(df['pl_bmasse'])
        df['radius_residual'] = df['pl_rade'] - df['predicted_radius']

        # Classify planets based on theoretical models
        df['planet_class'] = df.apply(self.classify_planet, axis=1)

        # Calculate habitable zone positions
        df['hz_inner'] = 0.75 * (df['st_teff']/5777)**2
        df['hz_outer'] = 1.77 * (df['st_teff']/5777)**2
        df['in_hz'] = (df['pl_orbsmax'] >= df['hz_inner']) & (df['pl_orbsmax'] <= df['hz_outer'])

        self.analyzed_data = df
        print("âœ“ Discoveries matched to theoretical models")
        return df

    def predict_radius(self, mass):
        """Predict radius based on theoretical mass-radius relations"""
        mass = np.array(mass)
        radius = np.zeros_like(mass)

        # Apply different relations based on mass
        # Rocky planets
        mask1 = mass < 10
        radius[mask1] = mass[mask1]**0.3 * 1.1

        # Water worlds/Neptunes
        mask2 = (mass >= 10) & (mass < 100)
        radius[mask2] = mass[mask2]**0.5 * 0.8

        # Gas giants
        mask3 = mass >= 100
        radius[mask3] = 10 * mass[mask3]**0.1 / (1 + mass[mask3]/1000)

        return radius

    def classify_planet(self, row):
        """Classify planet based on theoretical models"""
        try:
            mass = row['pl_bmasse']
            radius = row['pl_rade']
            density = mass / radius**3 if radius > 0 else 0

            if mass < 2:
                return 'Super-Earth'
            elif mass < 10:
                return 'Mini-Neptune' if density < 3 else 'Rocky'
            elif mass < 100:
                return 'Neptune-like'
            else:
                return 'Gas Giant'
        except:
            return 'Unknown'

    def statistical_analysis(self):
        """Perform statistical analysis comparing theory and observations"""
        print("Performing statistical analysis...")

        df = self.analyzed_data

        stats_results = {
            'correlation_mass_radius': {
                'pearson': stats.pearsonr(df['pl_bmasse'], df['pl_rade'])[0],
                'spearman': stats.spearmanr(df['pl_bmasse'], df['pl_rade'])[0]
            },
            'radius_predictions': {
                'mae': np.mean(np.abs(df['radius_residual'])),
                'rmse': np.sqrt(np.mean(df['radius_residual']**2)),
                'r2': 1 - np.var(df['radius_residual']) / np.var(df['pl_rade'])
            },
            'discovery_method_stats': df.groupby('discoverymethod').agg({
                'pl_bmasse': ['mean', 'std', 'count'],
                'pl_rade': ['mean', 'std']
            }).to_dict(),
            'habitable_zone_stats': {
                'total_in_hz': df['in_hz'].sum(),
                'percent_in_hz': df['in_hz'].mean() * 100
            }
        }

        self.statistical_results = stats_results
        print("âœ“ Statistical analysis complete")
        return stats_results

    def plot_comparative_analysis(self):
        """Create comparative plots between theory and observations"""
        print("Generating comparative analysis plots...")

        fig, axes = plt.subplots(2, 3, figsize=(18, 12))
        fig.suptitle('Exoplanet Discoveries vs Theoretical Predictions', fontsize=16, y=1.02)

        df = self.analyzed_data

        # 1. Mass-Radius Comparison
        ax = axes[0, 0]
        scatter = ax.scatter(df['pl_bmasse'], df['pl_rade'],
                           c=df['st_teff'], cmap='viridis',
                           alpha=0.6, s=20)
        ax.set_xscale('log')
        ax.set_yscale('log')
        ax.set_xlabel('Mass (Earth masses)')
        ax.set_ylabel('Radius (Earth radii)')
        ax.set_title('Observed Mass-Radius Relation')
        plt.colorbar(scatter, ax=ax, label='Stellar Temp (K)')

        # Overlay theoretical curves
        theory = self.theoretical_data['mass_radius_relation']
        for key, model in theory.items():
            ax.plot(model['mass'], model['radius'], '--', alpha=0.7, label=key)
        ax.legend(fontsize=8)

        # 2. Radius Residuals
        ax = axes[0, 1]
        ax.hist(df['radius_residual'], bins=50, alpha=0.7, density=True)
        ax.axvline(0, color='red', linestyle='--', alpha=0.5)
        ax.set_xlabel('Radius Residual (Observed - Predicted)')
        ax.set_ylabel('Density')
        ax.set_title('Radius Prediction Errors')

        # 3. Habitable Zone Analysis
        ax = axes[0, 2]
        hz_data = self.theoretical_data['habitable_zone']
        ax.fill_between(hz_data['star_temp'],
                       hz_data['conservative']['inner'],
                       hz_data['conservative']['outer'],
                       alpha=0.3, label='Conservative HZ')
        ax.fill_between(hz_data['star_temp'],
                       hz_data['optimistic']['inner'],
                       hz_data['optimistic']['outer'],
                       alpha=0.3, label='Optimistic HZ')
        ax.scatter(df['st_teff'], df['pl_orbsmax'],
                  c=df['in_hz'].map({True: 'green', False: 'gray'}),
                  alpha=0.6, s=20)
        ax.set_xlabel('Stellar Temperature (K)')
        ax.set_ylabel('Orbital Distance (AU)')
        ax.set_title('Planets in Habitable Zone')
        ax.legend()

        # 4. Discovery Method Comparison
        ax = axes[1, 0]
        method_counts = df['discoverymethod'].value_counts()
        ax.pie(method_counts.values, labels=method_counts.index, autopct='%1.1f%%')
        ax.set_title('Exoplanet Discovery Methods')

        # 5. Planet Classification
        ax = axes[1, 1]
        class_counts = df['planet_class'].value_counts()
        bars = ax.bar(range(len(class_counts)), class_counts.values)
        ax.set_xticks(range(len(class_counts)))
        ax.set_xticklabels(class_counts.index, rotation=45, ha='right')
        ax.set_ylabel('Number of Planets')
        ax.set_title('Planet Classification')

        # Add counts on bars
        for bar, count in zip(bars, class_counts.values):
            height = bar.get_height()
            ax.text(bar.get_x() + bar.get_width()/2., height,
                   f'{count}', ha='center', va='bottom')

        # 6. Migration Models
        ax = axes[1, 2]
        migration = self.theoretical_data['migration_models']
        ax.plot(migration['type_I']['planet_mass'],
                migration['type_I']['migration_rate'],
                label='Type I Migration')
        ax.plot(migration['type_II']['disk_mass']*1000,
                migration['type_II']['migration_rate'],
                label='Type II Migration')
        ax.set_xlabel('Mass (Earth masses)')
        ax.set_ylabel('Migration Rate (AU/year)')
        ax.set_xscale('log')
        ax.set_yscale('log')
        ax.set_title('Planetary Migration Models')
        ax.legend()

        plt.tight_layout()
        plt.savefig('exoplanet_theoretical_comparison.png', dpi=150, bbox_inches='tight')
        plt.show()

        print("âœ“ Comparative analysis plots generated")

    def advanced_analysis(self):
        """Perform advanced theoretical analysis"""
        print("Performing advanced theoretical analysis...")

        df = self.analyzed_data

        # 1. Calculate Hill radii
        df['hill_radius'] = df['pl_orbsmax'] * (df['pl_bmasse']/(3*317.8))**(1/3)

        # 2. Estimate equilibrium temperatures
        df['eq_temp'] = 280 * df['st_teff']/5777 * (df['pl_orbsmax'])**(-0.5)

        # 3. Calculate Roche limits
        df['roche_limit'] = 2.44 * df['pl_rade'] * (317.8/df['pl_bmasse'])**(1/3)

        # 4. Estimate atmospheric scale heights
        df['scale_height'] = 8.31 * df['eq_temp'] / (0.02896 * 9.81) / 1000  # km

        advanced_results = {
            'hill_radius_stats': df['hill_radius'].describe().to_dict(),
            'eq_temp_stats': df['eq_temp'].describe().to_dict(),
            'scale_height_stats': df['scale_height'].describe().to_dict(),
            'planet_density': (df['pl_bmasse'] / df['pl_rade']**3).describe().to_dict()
        }

        self.advanced_results = advanced_results
        print("âœ“ Advanced analysis complete")
        return advanced_results

    def generate_report(self):
        """Generate a comprehensive analysis report"""
        print("\n" + "="*60)
        print("EXOPLANET THEORETICAL ANALYSIS REPORT")
        print("="*60)

        df = self.analyzed_data

        print(f"\nðŸ“Š DATA SUMMARY:")
        print(f"   Total planets analyzed: {len(df)}")
        print(f"   Data range: Mass = {df['pl_bmasse'].min():.2f} - {df['pl_bmasse'].max():.2f} MâŠ•")
        print(f"               Radius = {df['pl_rade'].min():.2f} - {df['pl_rade'].max():.2f} RâŠ•")

        print(f"\nðŸ”­ DISCOVERY METHODS:")
        for method, count in df['discoverymethod'].value_counts().items():
            print(f"   {method}: {count} planets ({count/len(df)*100:.1f}%)")

        print(f"\nðŸª PLANET CLASSIFICATION:")
        for pclass, count in df['planet_class'].value_counts().items():
            print(f"   {pclass}: {count} planets")

        stats = self.statistical_results
        print(f"\nðŸ“ˆ STATISTICAL ANALYSIS:")
        print(f"   Mass-Radius correlation: r = {stats['correlation_mass_radius']['pearson']:.3f}")
        print(f"   Radius prediction RMSE: {stats['radius_predictions']['rmse']:.3f} RâŠ•")
        print(f"   Planets in habitable zone: {stats['habitable_zone_stats']['total_in_hz']} "
              f"({stats['habitable_zone_stats']['percent_in_hz']:.1f}%)")

        print(f"\nâš›ï¸  THEORETICAL INSIGHTS:")
        print(f"   Core accretion efficiency range: "
              f"{self.theoretical_data['formation_models']['core_accretion']['efficiency'].min():.1f}-"
              f"{self.theoretical_data['formation_models']['core_accretion']['efficiency'].max():.1f}")
        print(f"   Disk instability temperature range: "
              f"{self.theoretical_data['formation_models']['disk_instability']['temperature'].min()}-"
              f"{self.theoretical_data['formation_models']['disk_instability']['temperature'].max()} K")

        if hasattr(self, 'advanced_results'):
            print(f"\nðŸ”¬ ADVANCED METRICS:")
            print(f"   Mean equilibrium temperature: {self.advanced_results['eq_temp_stats']['mean']:.1f} K")
            print(f"   Mean atmospheric scale height: {self.advanced_results['scale_height_stats']['mean']:.1f} km")

        print("\n" + "="*60)
        print("Analysis complete! Check generated plots for visual insights.")
        print("="*60 + "\n")

# @title Main Execution Cell
def main():
    """Main execution function"""
    print("ðŸš€ Starting Exoplanet Theoretical Analysis...\n")

    # Initialize the analysis system
    analysis = ExoplanetTheoreticalAnalysis()

    # Clean the data
    analysis.clean_data()

    # Match discoveries to theoretical models
    analysis.match_discoveries_to_theory()

    # Perform statistical analysis
    analysis.statistical_analysis()

    # Perform advanced analysis
    analysis.advanced_analysis()

    # Generate plots
    analysis.plot_comparative_analysis()

    # Generate report
    analysis.generate_report()

    # Save results to files
    print("\nðŸ’¾ Saving results to files...")
    analysis.analyzed_data.to_csv('exoplanet_theoretical_analysis.csv', index=False)

    with open('analysis_summary.txt', 'w') as f:
        f.write(str(analysis.statistical_results))

    print("âœ“ Results saved to 'exoplanet_theoretical_analysis.csv' and 'analysis_summary.txt'")

    return analysis

# @title Run Analysis
if __name__ == "__main__":
    analysis_results = main()

# @title Additional Analysis Functions
def plot_formation_models():
    """Plot planet formation models from theoretical data"""
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))

    theory = TheoreticalAstrophysicsData().theoretical_data

    # Core Accretion Model
    ax = axes[0]
    ca = theory['formation_models']['core_accretion']
    ax.plot(ca['timescale'], ca['mass_limit'], 'o-', linewidth=2)
    ax.set_xlabel('Formation Timescale (years)')
    ax.set_ylabel('Planet Mass Limit (Earth masses)')
    ax.set_xscale('log')
    ax.set_title('Core Accretion Model')
    ax.grid(True, alpha=0.3)

    # Disk Instability Model
    ax = axes[1]
    di = theory['formation_models']['disk_instability']
    scatter = ax.scatter(di['radius_au'], di['mass_jupiter'],
                        c=di['temperature'], s=200, cmap='hot')
    ax.set_xlabel('Formation Radius (AU)')
    ax.set_ylabel('Planet Mass (Jupiter masses)')
    ax.set_title('Disk Instability Model')
    plt.colorbar(scatter, ax=ax, label='Formation Temperature (K)')
    ax.grid(True, alpha=0.3)

    plt.suptitle('Planet Formation Theoretical Models', fontsize=14)
    plt.tight_layout()
    plt.show()

def interactive_mass_radius_explorer():
    """Interactive exploration of mass-radius relationships"""
    from ipywidgets import interact, FloatSlider

    theory = TheoreticalAstrophysicsData().theoretical_data

    def plot_interactive(composition=0.5, core_mass=5, envelope_fraction=0.1):
        """Interactive plot of mass-radius relationship"""
        fig, ax = plt.subplots(figsize=(10, 6))

        # Plot theoretical curves
        for key, model in theory['mass_radius_relation'].items():
            ax.plot(model['mass'], model['radius'], '--', alpha=0.5, label=key)

        # Calculate custom planet
        total_mass = core_mass / (1 - envelope_fraction)
        if composition < 0.33:
            # Rocky
            radius = total_mass**0.3 * 1.1
        elif composition < 0.66:
            # Water world
            radius = total_mass**0.5 * 0.8
        else:
            # Gas-rich
            radius = 10 * total_mass**0.1 / (1 + total_mass/1000)

        ax.plot(total_mass, radius, 'ro', markersize=15, label='Your Planet')

        ax.set_xlabel('Mass (Earth masses)')
        ax.set_ylabel('Radius (Earth radii)')
        ax.set_xscale('log')
        ax.set_yscale('log')
        ax.set_title(f'Custom Planet: Mass={total_mass:.1f} MâŠ•, Radius={radius:.1f} RâŠ•')
        ax.legend()
        ax.grid(True, alpha=0.3)
        plt.show()

    interact(plot_interactive,
             composition=FloatSlider(value=0.5, min=0, max=1, step=0.1,
                                   description='Composition:'),
             core_mass=FloatSlider(value=5, min=0.1, max=50, step=0.1,
                                 description='Core Mass (MâŠ•):'),
             envelope_fraction=FloatSlider(value=0.1, min=0, max=0.9, step=0.05,
                                         description='Envelope Fraction:'))

# @title Run Additional Analyses
print("Running additional analyses...\n")

# Plot formation models
plot_formation_models()

print("\nTo run interactive explorer, uncomment and run:")
print("# interactive_mass_radius_explorer()")

# @title Save Final Notebook State
import pickle

def save_analysis_state(analysis, filename='exoplanet_analysis_state.pkl'):
    """Save the analysis state for future use"""
    with open(filename, 'wb') as f:
        pickle.dump({
            'analyzed_data': analysis.analyzed_data,
            'statistical_results': analysis.statistical_results,
            'advanced_results': analysis.advanced_results,
            'theoretical_data': analysis.theoretical_data
        }, f)
    print(f"âœ“ Analysis state saved to {filename}")

# Save the state
save_analysis_state(analysis_results)

print("\nâœ… EXOPLANET THEORETICAL ANALYSIS COMPLETE!")
print("ðŸ“ Generated files:")
print("   - exoplanet_theoretical_analysis.csv (cleaned data with predictions)")
print("   - analysis_summary.txt (statistical results)")
print("   - exoplanet_theoretical_comparison.png (analysis plots)")
print("   - exoplanet_analysis_state.pkl (complete analysis state)")
print("\nðŸŽ¯ Next steps:")
print("   1. Upload your actual exoplanet_discovery_v6_3.txt file")
print("   2. Extract data from extract_thesis5.pdf using PyPDF2")
print("   3. Update the data loading functions with your real data")
print("   4. Run the analysis again with complete datasets")
```
